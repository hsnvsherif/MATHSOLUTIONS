**Solving the System of Linear Equations**

Given:

$$ 
\begin{cases}
2x - 3y + 4z + 2t = 2 \\
3x + 2y - 5z + 3t = 3 \\
4x - 3y + 2z - 5t = 4 \\
5x + 4y - 3z + 2t = 5
\end{cases}
$$

**Using Gaussian Elimination:**

1. **Augmented Matrix:** 
   $$ 
   \left[
   \begin{array}{cccc|c}
   2 & -3 & 4 & 2 & 2 \\
   3 & 2 & -5 & 3 & 3 \\
   4 & -3 & 2 & -5 & 4 \\
   5 & 4 & -3 & 2 & 5
   \end{array}
   \right]
   $$

2. **Row Operations:** 
   Perform row operations (e.g., scaling, adding/subtracting rows) to obtain row-echelon form.

3. **Back Substitution:** 
   Solve for variables starting from the last row.

**Solution:**

- From row 3: \( z = 0 \)
- From row 2: \( y = 0 \)
- From row 1: \( x = 1 \)

Thus, the solution is:

$$ 
x = 1, \quad y = 0, \quad z = 0, \quad t = 0
$$

## Exercises for Students

$$
\left\{\begin{array}{c}{x+2 y-2 z=4} \\ {2 x+y+z=0} \\ {3 x+2 y+z=1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y+z-t=2} \\ {2 x+y+z=3} \\ {-x+z-t=0} \\ {3 x+2 y-z+2 t=-1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y-z-t=0} \\ {2 x+3 y-2 z+t=4} \\ {3 x+5 z=0} \\ {-x+y-3 z+2 t=3}\end{array}\right.
$$

GAUSS ELIMINATION METHOD 

### **System 1**

The system of equations is:

$$
\begin{aligned}
x + 2y - 2z &= 4, \\
2x + y + z &= 0, \\
3x + 2y + z &= 1.
\end{aligned}
$$

This can be written as the augmented matrix:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
2 && 1 &&  1 &&  0 \\
3 && 2 &&  1 &&  1
\end{bmatrix}.
$$

#### Step 1: Eliminate \(x\) from the second and third rows.

Subtract \(2 \times R_1\) from \(R_2\): \(R_2 \to R_2 - 2R_1\),  
and subtract \(3 \times R_1\) from \(R_3\): \(R_3 \to R_3 - 3R_1\).

The resulting matrix is:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
0 && -3 &&  5 && -8 \\
0 && -4 &&  7 && -11
\end{bmatrix}.
$$

#### Step 2: Eliminate \(y\) from the third row.

Divide \(R_2\) by \(-3\): \(R_2 \to \frac{R_2}{-3}\),  
then use \(R_2\) to eliminate \(y\) from \(R_3\): \(R_3 \to R_3 - (-4) \cdot R_2\).

The resulting matrix is:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
0 && 1 && -\frac{5}{3} && \frac{8}{3} \\
0 && 0 && \frac{2}{3} && \frac{2}{3}
\end{bmatrix}.
$$

#### Step 3: Back-substitution.

From \(R_3\):

$$
\frac{2}{3}z = \frac{2}{3} \implies z = 1.
$$

From \(R_2\):

$$
y - \frac{5}{3}(1) = \frac{8}{3} \implies y = \frac{13}{3}.
$$

From \(R_1\):

$$
x + 2\left(\frac{13}{3}\right) - 2(1) = 4 \implies x = -2.
$$

#### Final Solution:

$$
x = -2, \quad y = \frac{13}{3}, \quad z = 1.
$$

$$
\left\{\begin{array}{c}{x+y+z-t=2} \\ {2 x+y+z=3} \\ {-x+z-t=0} \\ {3 x+2 y-z+2 t=-1}\end{array}\right.\quad
$$

**System:**

$$
\left\{\begin{array}{c}{x+y+z-t=2} \\ {2 x+y+z=3} \\ {-x+z-t=0} \\ {3 x+2 y-z+2 t=-1}\end{array}\right.\quad
$$

**1. Augmented Matrix:**

Represent the system as an augmented matrix:

$$
\left[\begin{array}{cccc|c}
1 & 1 & 1 & -1 & 2 \\
2 & 1 & 1 & 0 & 3 \\
-1 & 0 & 1 & -1 & 0 \\
3 & 2 & -1 & 2 & -1
\end{array}\right]
$$

**2. Row Operations:**

*   **R2 = R2 - 2*R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & -1 & -1 & 2 & -1 \\
    -1 & 0 & 1 & -1 & 0 \\
    3 & 2 & -1 & 2 & -1
    \end{array}\right]
    $$
*   **R3 = R3 + R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & -1 & -1 & 2 & -1 \\
    0 & 1 & 2 & -2 & 2 \\
    3 & 2 & -1 & 2 & -1
    \end{array}\right]
    $$
*   **R4 = R4 - 3*R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & -1 & -1 & 2 & -1 \\
    0 & 1 & 2 & -2 & 2 \\
    0 & -1 & -4 & 5 & -7
    \end{array}\right]
    $$
*   **R2 = -1*R2:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & 1 & 1 & -2 & 1 \\
    0 & 1 & 2 & -2 & 2 \\
    0 & -1 & -4 & 5 & -7
    \end{array}\right]
    $$
*   **R3 = R3 - R2:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & 1 & 1 & -2 & 1 \\
    0 & 0 & 1 & 0 & 1 \\
    0 & -1 & -4 & 5 & -7
    \end{array}\right]
    $$
*   **R4 = R4 + R2:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & 1 & 1 & -2 & 1 \\
    0 & 0 & 1 & 0 & 1 \\
    0 & 0 & -3 & 3 & -6
    \end{array}\right]
    $$
*   **R4 = R4 + 3*R3:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & 1 & 1 & -2 & 1 \\
    0 & 0 & 1 & 0 & 1 \\
    0 & 0 & 0 & 3 & -3
    \end{array}\right]
    $$
*   **R4 = (1/3)*R4:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & 1 & -1 & 2 \\
    0 & 1 & 1 & -2 & 1 \\
    0 & 0 & 1 & 0 & 1 \\
    0 & 0 & 0 & 1 & -1
    \end{array}\right]
    $$

**3. Back-Substitution:**

*   From R4: t = -1
*   From R3: z = 1
*   From R2: y + 1 - 2*(-1) = 1 => y = -2
*   From R1: x + (-2) + 1 - (-1) = 2 => x = 2

**Solution:**

Therefore, the solution to the system is:

x = 2, y = -2, z = 1, t = -1

**System 3:**

$$
\left\{\begin{array}{c}{x+y-z-t=0} \\ {2 x+3 y-2 z+t=4} \\ {3 x+5 z=0} \\ {-x+y-3 z+2 t=3}\end{array}\right.\quad
$$

**1. Augmented Matrix:**

$$
\left[\begin{array}{cccc|c}
1 & 1 & -1 & -1 & 0 \\
2 & 3 & -2 & 1 & 4 \\
3 & 0 & 5 & 0 & 0 \\
-1 & 1 & -3 & 2 & 3
\end{array}\right]
$$

**2. Row Operations:**

*   **R2 = R2 - 2*R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    3 & 0 & 5 & 0 & 0 \\
    -1 & 1 & -3 & 2 & 3
    \end{array}\right]
    $$
*   **R3 = R3 - 3*R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    0 & -3 & 8 & 3 & 0 \\
    -1 & 1 & -3 & 2 & 3
    \end{array}\right]
    $$
*   **R4 = R4 + R1:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    0 & -3 & 8 & 3 & 0 \\
    0 & 2 & -4 & 1 & 3
    \end{array}\right]
    $$
*   **R3 = R3 + 3*R2:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    0 & 0 & 8 & 12 & 12 \\
    0 & 2 & -4 & 1 & 3
    \end{array}\right]
    $$
*   **R4 = R4 - 2*R2:** 
    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    0 & 0 & 8 & 12 & 12 \\

    0 & 1 & 0 & 3 & 4 \\
    0 & 0 & 1 & 3/2 & 3/2 \\
    0 & 0 & -4 & -5 & -5
    \end{array}\right]
    $$

    $$
    \left[\begin{array}{cccc|c}
    1 & 1 & -1 & -1 & 0 \\
    0 & 1 & 0 & 3 & 4 \\
    0 & 0 & 1 & 3/2 & 3/2 \\
    0 & 0 & 0 & 1 & 1
    \end{array}\right]
    $$

**3. Back-Substitution:**

*   From R4: t = 1
*   From R3: z + (3/2)*(1) = 3/2 => z = 0
*   From R2: y + 3*(1) = 4 => y = 1
*   From R1: x + 1 - 0 - 1 = 0 => x = 0

**Solution:**

Therefore, the solution to the system is:

x = 0, y = 1, z = 0, t = 1

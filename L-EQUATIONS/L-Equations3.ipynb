## Exercises for Students

$$
\left\{\begin{array}{c}{x+2 y-2 z=4} \\ {2 x+y+z=0} \\ {3 x+2 y+z=1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y+z-t=2} \\ {2 x+y+z=3} \\ {-x+z-t=0} \\ {3 x+2 y-z+2 t=-1}\end{array}\right.\quad
\left\{\begin{array}{c}{x+y-z-t=0} \\ {2 x+3 y-2 z+t=4} \\ {3 x+5 z=0} \\ {-x+y-3 z+2 t=3}\end{array}\right.
$$

GAUSS ELIMINATION METHOD 

### **System 1**

The system of equations is:

$$
\begin{aligned}
x + 2y - 2z &= 4, \\
2x + y + z &= 0, \\
3x + 2y + z &= 1.
\end{aligned}
$$

This can be written as the augmented matrix:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
2 && 1 &&  1 &&  0 \\
3 && 2 &&  1 &&  1
\end{bmatrix}.
$$

#### Step 1: Eliminate \(x\) from the second and third rows.

Subtract \(2 \times R_1\) from \(R_2\): \(R_2 \to R_2 - 2R_1\),  
and subtract \(3 \times R_1\) from \(R_3\): \(R_3 \to R_3 - 3R_1\).

The resulting matrix is:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
0 && -3 &&  5 && -8 \\
0 && -4 &&  7 && -11
\end{bmatrix}.
$$

#### Step 2: Eliminate \(y\) from the third row.

Divide \(R_2\) by \(-3\): \(R_2 \to \frac{R_2}{-3}\),  
then use \(R_2\) to eliminate \(y\) from \(R_3\): \(R_3 \to R_3 - (-4) \cdot R_2\).

The resulting matrix is:

$$
\begin{bmatrix}
1 && 2 && -2 &&  4 \\
0 && 1 && -\frac{5}{3} && \frac{8}{3} \\
0 && 0 && \frac{2}{3} && \frac{2}{3}
\end{bmatrix}.
$$

#### Step 3: Back-substitution.

From \(R_3\):

$$
\frac{2}{3}z = \frac{2}{3} \implies z = 1.
$$

From \(R_2\):

$$
y - \frac{5}{3}(1) = \frac{8}{3} \implies y = \frac{13}{3}.
$$

From \(R_1\):

$$
x + 2\left(\frac{13}{3}\right) - 2(1) = 4 \implies x = -2.
$$

#### Final Solution:

$$
x = -2, \quad y = \frac{13}{3}, \quad z = 1.
$$

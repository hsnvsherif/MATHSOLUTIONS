
1. Solve the system of equations:
   $$
   \begin{cases}
   2x_1 - 3x_2 = 7 \\
   3x_1 + 5x_2 = 2
   \end{cases}
   $$
   

---


The system of equations is:  
$$
\begin{cases}
2x_1 - 3x_2 = 7 \\
3x_1 + 5x_2 = 2
\end{cases}
$$

1. Write the system in matrix form:
   $$
   \mathbf{A} \mathbf{x} = \mathbf{b}, \quad \text{where} \quad 
   \mathbf{A} = \begin{bmatrix} 2 & -3 \\ 3 & 5 \end{bmatrix}, \quad 
   \mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}, \quad 
   \mathbf{b} = \begin{bmatrix} 7 \\ 2 \end{bmatrix}.
   $$

2. Compute the determinant of the coefficient matrix :
   $$
   \Delta = \det(\mathbf{A}) = \begin{vmatrix} 2 & -3 \\ 3 & 5 \end{vmatrix} = (2)(5) - (3)(-3) = 10 + 9 = 19.
   $$

3. Compute \(\Delta_{x_1}\), the determinant of the matrix formed by replacing the first column of \(\mathbf{A}\) with \(\mathbf{b}\):
   $$
   \Delta_{x_1} = \begin{vmatrix} 7 & -3 \\ 2 & 5 \end{vmatrix} = (7)(5) - (2)(-3) = 35 + 6 = 41.
   $$

4. Compute \(\Delta_{x_2}\), the determinant of the matrix formed by replacing the second column of \(\mathbf{A}\) with \(\mathbf{b}\):
   $$
   \Delta_{x_2} = \begin{vmatrix} 2 & 7 \\ 3 & 2 \end{vmatrix} = (2)(2) - (3)(7) = 4 - 21 = -17.
   $$

5. Solve for \(x_1\) and \(x_2\):
   $$
   x_1 = \frac{\Delta_{x_1}}{\Delta} = \frac{41}{19}, \quad 
   x_2 = \frac{\Delta_{x_2}}{\Delta} = \frac{-17}{19}.
   $$

### Final Solution:
$$
x_1 = \frac{41}{19}, \quad x_2 = \frac{-17}{19}.
$$

   $$
   \Delta = \det(\mathbf{A}) =

   $$
   \Delta = 2 \begin{vmatrix} 
   -1 & 2 \\

   \end{vmatrix} 
   - 1 \begin{vmatrix} 
   1 & 2 \\ 
   3 & -2 
   \end{vmatrix}

   1 & -1 \\ 
   3 & 0

   $$

   -1 & 2 \\

   \begin{vmatrix}

4. Why can't the following system of equations be solved using Cramer's rule?
$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 3 \\
4x_1 + 5x_2 + 6x_3 = 2 \\
7x_1 + 8x_2 + 9x_3 = 1
\end{cases}
$$



**To determine why this system of equations cannot be solved using Cramer's Rule, let's examine the necessary condition for applying the rule.**

### Cramer's Rule Condition

Cramer's Rule requires that the determinant of the coefficient matrix (\(\mathbf{A}\)) is **nonzero**. If $\det(\mathbf{A}) = 0$, the system of equations either has:
1. **No solution** (inconsistent system), or  
2. **Infinitely many solutions** (dependent system).

### Coefficient Matrix

The coefficient matrix \(\mathbf{A}\) for the given system is:

$$
\mathbf{A} = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}.
$$

### Determinant of \(\mathbf{A}\)

The determinant of \(\mathbf{A}\) is computed as:

$$
\det(\mathbf{A}) = 1 \cdot \begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} - 2 \cdot \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} + 3 \cdot \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix}.
$$

Expanding the minors:

$$
\begin{aligned}
\det(\mathbf{A}) &= 1 \cdot (5 \cdot 9 - 6 \cdot 8) - 2 \cdot (4 \cdot 9 - 6 \cdot 7) + 3 \cdot (4 \cdot 8 - 5 \cdot 7) \\
&= 1 \cdot (45 - 48) - 2 \cdot (36 - 42) + 3 \cdot (32 - 35) \\
&= 1 \cdot (-3) - 2 \cdot (-6) + 3 \cdot (-3) \\
&= -3 + 12 - 9 \\
&= 0.
\end{aligned}
$$

### Conclusion

Since $\det(\mathbf{A}) = 0$, the coefficient matrix is **singular**, meaning the system cannot be solved using Cramer's Rule. This occurs because the rows of \(\mathbf{A}\) are linearly dependent (one row can be expressed as a combination of the others), leading to:
1. **No solution** if the system is inconsistent, or  
2. **Infinitely many solutions** if the system is dependent.

### Final Answer

The given system cannot be solved using Cramer's Rule because $\det(\mathbf{A}) = 0$.

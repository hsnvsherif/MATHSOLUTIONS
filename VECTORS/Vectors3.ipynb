- Divide the line segment connecting points $A(-1, 2)$ and $B(3, -2)$ in the ratio $1:3$. Illustrate the result on a graph.

We are given points \( A(-1, 2) \) and \( B(3, -2) \), and we want to divide the line segment \( AB \) in the ratio \( 1:3 \).

The formula for dividing a line segment in the ratio \( m:n \) is:

$$
P = \left( \frac{n \cdot x_1 + m \cdot x_2}{m + n}, \frac{n \cdot y_1 + m \cdot y_2}{m + n} \right)
$$

Substituting the values \( m = 1 \), \( n = 3 \), and the coordinates of \( A(-1, 2) \) and \( B(3, -2) \):

$$
P = \left( \frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3}, \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3} \right)
$$

Simplifying:

$$
P = \left( \frac{3 - 3}{4}, \frac{-2 + 6}{4} \right)
$$

$$
P = \left( 0, 1 \right)
$$

Thus, the point \( P(0, 1) \) divides the segment \( AB \) in the ratio \( 1:3 \).

   * $$ \overrightarrow{AB} = B - A = (3, -2) - (-1, 2) = (4, -4) $$

* **Determine the position vector of the division point:**
   * Let P be the point that divides AB in the ratio 1:3. 
   * The position vector of P is given by:
      * $$ \overrightarrow{OP} = \overrightarrow{OA} + \frac{1}{1+3} \overrightarrow{AB} $$

      * $$ \overrightarrow{OP} = \langle -1, 2 \rangle + \langle 1, -1 \rangle $$
      * $$ \overrightarrow{OP} = \langle 0, 1 \rangle $$

* **Find the coordinates of P:**
   * The coordinates of P are (0, 1).

 Project vector $\mathbf{a} = (2,3)$ onto vector $\mathbf{b} = (1, 1)$. Illustrate the result on a graph.
Given the vectors \( \mathbf{a} = (2, 3) \) and \( \mathbf{b} = (1, 1) \), we want to find the projection of \( \mathbf{a} \) onto \( \mathbf{b} \).

The formula for the projection of \( \mathbf{a} \) onto \( \mathbf{b} \) is:

$$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$

First, we compute the dot products:

$$
\mathbf{a} \cdot \mathbf{b} = 2 \cdot 1 + 3 \cdot 1 = 2 + 3 = 5
$$

$$
\mathbf{b} \cdot \mathbf{b} = 1 \cdot 1 + 1 \cdot 1 = 1 + 1 = 2
$$

Now, we compute the projection:

$$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \mathbf{b} = \frac{5}{2} \cdot (1, 1) = \left( \frac{5}{2}, \frac{5}{2} \right)
$$

$$
\text{proj}_{\mathbf{b}}(\mathbf{a}) = \left\langle \frac{5}{2}, \frac{5}{2} \right\rangle
$$
$$
\left( \frac{5}{2}, \frac{5}{2} \right)
$$

The formula for the projection of \( \mathbf{b} \) onto \( \mathbf{a} \) is:

$$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$$

First, we compute the dot products:

$$
\mathbf{b} \cdot \mathbf{a} = 1 \cdot 2 + 1 \cdot 3 = 2 + 3 = 5
$$

$$
\mathbf{a} \cdot \mathbf{a} = 2 \cdot 2 + 3 \cdot 3 = 4 + 9 = 13
$$

Now, we compute the projection:

$$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{5}{13} \mathbf{a} = \frac{5}{13} \cdot (2, 3) = \left( \frac{10}{13}, \frac{15}{13} \right)
$$
Thus, the projection of the vector \( \mathbf{b} = (1, 1) \) onto the vector \( \mathbf{a} = (2, 3) \) is:

**(10/13, 15/13)**.

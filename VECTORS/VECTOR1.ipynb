
A. By what number should vector $\mathbf{a} = [3, 4]$ be multiplied so that its length is equal to 1?

---

**1. Calculate the length of vector a**

Given the vector $\mathbf{a} = [3, 4]$, its length (or magnitude) is calculated as:

\begin{equation*}
|\mathbf{a}| = \sqrt{a_1^2 + a_2^2} = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
\end{equation*}

**2. Determine the scalar to multiply vector a**

To find the scalar, let's denote it as 'k'. We need to find the value of 'k' such that the length of the scaled vector, $k\mathbf{a}$, is equal to 1. 

The length of the scaled vector is:

\begin{equation*}
|k\mathbf{a}| = |k| |\mathbf{a}| = |k| \cdot 5
\end{equation*}

We want:

\begin{equation*}
|k| \cdot 5 = 1
\end{equation*}

Therefore:

\begin{equation*}
|k| = \frac{1}{5} 
\end{equation*}

Since we need a unit vector, we can choose:

\begin{equation*}
k = \frac{1}{5} 
\end{equation*}

**Therefore:**

Vector $\mathbf{a} = [3, 4]$ should be multiplied by **1/5** to make its length equal to 1.

B. Calculate the length of vector $\mathbf{b} = [1, 1]$ and find the unit vector of this vector.

Given the vector $\mathbf{b} = [1, 1]$, its length (or magnitude) is calculated as:

\begin{equation*}
|\mathbf{b}| = \sqrt{b_1^2 + b_2^2} = \sqrt{1^2 + 1^2} = \sqrt{1 + 1} = \sqrt{2}
\end{equation*}

**2. Find the unit vector of b**

The unit vector of $\mathbf{b}$, denoted as $\hat{\mathbf{b}}$, is a vector in the same direction as $\mathbf{b}$ but with a magnitude of 1. It is calculated as:

\begin{equation*}
\hat{\mathbf{b}} = \frac{\mathbf{b}}{|\mathbf{b}|} = \frac{1}{\sqrt{2}} [1, 1] = \left[\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right]
\end{equation*}

**Therefore:**

* The length of vector $\mathbf{b}$ is $\sqrt{2}$.
* The unit vector of $\mathbf{b}$ is $\left[\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right]$.

**1. Calculate the length of vector c**

Given the vector $\mathbf{c} = [1, 2, 3]$, its length (or magnitude) is calculated as:

\begin{equation*}
|\mathbf{c}| = \sqrt{c_1^2 + c_2^2 + c_3^2} = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{1 + 4 + 9} = \sqrt{14}
\end{equation*}

**2. Find the unit vector of c**

The unit vector of $\mathbf{c}$, denoted as $\hat{\mathbf{c}}$, is a vector in the same direction as $\mathbf{c}$ but with a magnitude of 1. It is calculated as:

\begin{equation*}
\hat{\mathbf{c}} = \frac{\mathbf{c}}{|\mathbf{c}|} = \frac{1}{\sqrt{14}} [1, 2, 3] = \left[\frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}}\right]
\end{equation*}

**Therefore:**

* The length of vector $\mathbf{c}$ is $\sqrt{14}$.
* The unit vector of $\mathbf{c}$ is $\left[\frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}}\right]$.

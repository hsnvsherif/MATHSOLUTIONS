**B. Calculate the area of the triangle spanned by vectors $[2, 1]$ and $[-1, 1]$.**

To calculate the area of the triangle spanned by two vectors \( \mathbf{v}_1 = [2, 1] \) and \( \mathbf{v}_2 = [-1, 1] \), we use the formula for the area of a parallelogram, which is half the magnitude of the cross product of the vectors:

$$
A = \frac{1}{2} \left| \mathbf{v}_1 \times \mathbf{v}_2 \right|
$$

The cross product for 2D vectors is calculated as:

$$
\mathbf{v}_1 \times \mathbf{v}_2 = v_{1x} v_{2y} - v_{1y} v_{2x}
$$

Now, applying this to the given vectors:

$$
\mathbf{v}_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}, \quad \mathbf{v}_2 = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
$$

$$
\mathbf{v}_1 \times \mathbf{v}_2 = 2 \cdot 1 - 1 \cdot (-1) = 2 + 1 = 3
$$

Thus, the area of the triangle is:

$$
A = \frac{1}{2} \left| 3 \right| = \frac{3}{2}
$$

Therefore, the area of the triangle is \( \frac{3}{2} \).

**C. Calculate the volume of the parallelepiped spanned by vectors $[2, 1]$, $[-1, 1]$, and $[1, 2]$.**

The volume \( V \) of the parallelepiped spanned by the vectors \( \mathbf{v}_1 = [2, 1] \), \( \mathbf{v}_2 = [-1, 1] \), and \( \mathbf{v}_3 = [1, 2] \) is given by the absolute value of the scalar triple product:

$$
V = \left| \mathbf{v}_1 \cdot (\mathbf{v}_2 \times \mathbf{v}_3) \right|
$$

First, calculate the cross product \( \mathbf{v}_2 \times \mathbf{v}_3 \):

$$
\mathbf{v}_2 \times \mathbf{v}_3 = (-1)(2) - (1)(1) = -3
$$

Then, calculate the dot product:

$$
\mathbf{v}_1 \cdot (\mathbf{v}_2 \times \mathbf{v}_3) = 2(-3) + 1(0) = -6
$$

Finally, the volume is:

$$
V = \left| -6 \right| = 6
$$

**D. Check if vectors $[2, 1]$ and $[-1, 1]$ are perpendicular.**

To check if the vectors \( \mathbf{v}_1 = [2, 1] \) and \( \mathbf{v}_2 = [-1, 1] \) are perpendicular, we compute their dot product:

$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = (2)(-1) + (1)(1) = -2 + 1 = -1
$$

Since the dot product is not zero, the vectors are not perpendicular.

**E. Calculate the angle in degrees between vectors $[4,2,1]$ and $[1,3,2]$.**

To calculate the angle \( \theta \) between the vectors \( \mathbf{v}_1 = [4, 2, 1] \) and \( \mathbf{v}_2 = [1, 3, 2] \), we use the formula:

$$
\cos(\theta) = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{|\mathbf{v}_1| |\mathbf{v}_2|}
$$

### Step 1: Compute the dot product:
$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = (4)(1) + (2)(3) + (1)(2) = 4 + 6 + 2 = 12
$$

### Step 2: Compute the magnitudes:
$$
|\mathbf{v}_1| = \sqrt{4^2 + 2^2 + 1^2} = \sqrt{16 + 4 + 1} = \sqrt{21}
$$
$$
|\mathbf{v}_2| = \sqrt{1^2 + 3^2 + 2^2} = \sqrt{1 + 9 + 4} = \sqrt{14}
$$

### Step 3: Calculate the cosine of the angle:
$$
\cos(\theta) = \frac{12}{\sqrt{21} \cdot \sqrt{14}} = \frac{12}{\sqrt{294}}
$$

### Step 4: Calculate the angle:
$$
\theta = \cos^{-1}\left( \frac{12}{\sqrt{294}} \right)
$$

**F. Scalar Triple Product Identity**

Given three-dimensional vectors:

- \( \mathbf{a} = [a_x, a_y, a_z] \)
- \( \mathbf{b} = [b_x, b_y, b_z] \)
- \( \mathbf{c} = [c_x, c_y, c_z] \)

We need to prove that the following identity is satisfied:

$$
\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) = \mathbf{b} \cdot (\mathbf{c} \times \mathbf{a}) = \mathbf{c} \cdot (\mathbf{a} \times \mathbf{b})
$$

### Step 1: Compute \( \mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) \)

First, compute the cross product \( \mathbf{b} \times \mathbf{c} \). Using the determinant of a 3x3 matrix:

$$
\mathbf{b} \times \mathbf{c} = \left| \begin{matrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ b_x & b_y & b_z \\ c_x & c_y & c_z \end{matrix} \right|
$$

This gives:

$$
\mathbf{b} \times \mathbf{c} = \left( (b_y c_z - b_z c_y), (b_z c_x - b_x c_z), (b_x c_y - b_y c_x) \right)
$$

Now, compute the dot product \( \mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) \):

$$
\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) = a_x (b_y c_z - b_z c_y) + a_y (b_z c_x - b_x c_z) + a_z (b_x c_y - b_y c_x)
$$

### Step 2: Compute \( \mathbf{b} \cdot (\mathbf{c} \times \mathbf{a}) \)

Now compute the cross product \( \mathbf{c} \times \mathbf{a} \) and then the dot product \( \mathbf{b} \cdot (\mathbf{c} \times \mathbf{a}) \).

### Step 3: Compute \( \mathbf{c} \cdot (\mathbf{a} \times \mathbf{b}) \)

Similarly, compute the cross product \( \mathbf{a} \times \mathbf{b} \) and then the dot product \( \mathbf{c} \cdot (\mathbf{a} \times \mathbf{b}) \).

### Conclusion

Each of the three expressions will result in the same value, which proves the identity:

$$
\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) = \mathbf{b} \cdot (\mathbf{c} \times \mathbf{a}) = \mathbf{c} \cdot (\mathbf{a} \times \mathbf{b})
$$

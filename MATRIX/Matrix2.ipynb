## Exercises for Students

Calculate the determinants of the following matrices:

### 1. Matrix $\mathbf{A}$:
$$
\mathbf{A} =
\begin{pmatrix}
2 & 3 & 1 \\
1 & 4 & 0 \\
3 & 2 & 1
\end{pmatrix}
$$
**Solution:**
We can compute the determinant using cofactor expansion along the first row:
$$
\text{det}(\mathbf{A}) = 2 \cdot \begin{vmatrix} 4 & 0 \\ 2 & 1 \end{vmatrix} - 3 \cdot \begin{vmatrix} 1 & 0 \\ 3 & 1 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 4 \\ 3 & 2 \end{vmatrix}
$$
Calculating the minors:
$$
\begin{vmatrix} 4 & 0 \\ 2 & 1 \end{vmatrix} = 4 \cdot 1 - 0 \cdot 2 = 4
$$
$$
\begin{vmatrix} 1 & 0 \\ 3 & 1 \end{vmatrix} = 1 \cdot 1 - 0 \cdot 3 = 1
$$
$$
\begin{vmatrix} 1 & 4 \\ 3 & 2 \end{vmatrix} = 1 \cdot 2 - 4 \cdot 3 = 2 - 12 = -10
$$
Now, substituting back into the determinant:
$$
\text{det}(\mathbf{A}) = 2 \cdot 4 - 3 \cdot 1 + 1 \cdot (-10) = 8 - 3 - 10 = -5
$$

### 2. Matrix $\mathbf{B}$:
$$
\mathbf{B} =
\begin{pmatrix}
2 & 3 & 1 \\
1 & 4 & 0 \\
3 & 2 & 0
\end{pmatrix}
$$
**Solution:**
Using cofactor expansion along the first row:
$$
\text{det}(\mathbf{B}) = 2 \cdot \begin{vmatrix} 4 & 0 \\ 2 & 0 \end{vmatrix} - 3 \cdot \begin{vmatrix} 1 & 0 \\ 3 & 0 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 4 \\ 3 & 2 \end{vmatrix}
$$
Calculating the minors:
$$
\begin{vmatrix} 4 & 0 \\ 2 & 0 \end{vmatrix} = 4 \cdot 0 - 0 \cdot 2 = 0
$$
$$
\begin{vmatrix} 1 & 0 \\ 3 & 0 \end{vmatrix} = 1 \cdot 0 - 0 \cdot 3 = 0
$$
$$
\begin{vmatrix} 1 & 4 \\ 3 & 2 \end{vmatrix} = 1 \cdot 2 - 4 \cdot 3 = 2 - 12 = -10
$$
Now, substituting back into the determinant:
$$
\text{det}(\mathbf{B}) = 2 \cdot 0 - 3 \cdot 0 + 1 \cdot (-10) = 0 + 0 - 10 = -10
$$

### 3. Matrix $\mathbf{C}$:
$$
\mathbf{C} =
\begin{pmatrix}
2 & 3 & 1 & 4 \\
1 & 0 & 0 & 6 \\
3 & 2 & 1 & 5 \\
2 & 1 & 4 & 0
\end{pmatrix}
$$

**Solution:**
Using cofactor expansion along the first row:

$$
\text{det}(\mathbf{C}) = 2 \cdot \begin{vmatrix} 0 & 0 & 6 \\ 2 & 1 & 5 \\ 1 & 4 & 0 \end{vmatrix} - 3 \cdot \begin{vmatrix} 1 & 0 & 6 \\ 3 & 1 & 5 \\ 2 & 4 & 0 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 0 & 0 \\ 3 & 2 & 5 \\ 2 & 1 & 4 \end{vmatrix} - 4 \cdot \begin{vmatrix} 1 & 0 & 0 \\ 3 & 2 & 1 \\ 2 & 1 & 4 \end{vmatrix}
$$

Now, let's calculate each 3x3 determinant:

#### First determinant:
$$
\begin{vmatrix} 
0 & 0 & 6 \\
2 & 1 & 5 \\
1 & 4 & 0 
\end{vmatrix}
= 0 \cdot \begin{vmatrix} 1 & 5 \\ 4 & 0 \end{vmatrix} - 0 \cdot \begin{vmatrix} 2 & 5 \\ 1 & 0 \end{vmatrix} + 6 \cdot \begin{vmatrix} 2 & 1 \\ 1 & 4 \end{vmatrix}
$$

$$
= 6 \cdot \left( 2 \cdot 4 - 1 \cdot 1 \right) = 6 \cdot (8 - 1) = 6 \cdot 7 = 42
$$

#### Second determinant:
$$
\begin{vmatrix}
1 & 0 & 6 \\
3 & 1 & 5 \\
2 & 4 & 0
\end{vmatrix}
= 1 \cdot \begin{vmatrix} 1 & 5 \\ 4 & 0 \end{vmatrix} - 0 \cdot \begin{vmatrix} 3 & 5 \\ 2 & 0 \end{vmatrix} + 6 \cdot \begin{vmatrix} 3 & 1 \\ 2 & 4 \end{vmatrix}
$$

$$
= 1 \cdot \left( 1 \cdot 0 - 5 \cdot 4 \right) + 6 \cdot \left( 3 \cdot 4 - 1 \cdot 2 \right)
$$

$$
= 1 \cdot (-20) + 6 \cdot (12 - 2) = -20 + 6 \cdot 10 = -20 + 60 = 40
$$

#### Third determinant:
$$
\begin{vmatrix}
1 & 0 & 0 \\
3 & 2 & 5 \\
2 & 1 & 4
\end{vmatrix}
= 1 \cdot \begin{vmatrix} 2 & 5 \\ 1 & 4 \end{vmatrix} - 0 \cdot \begin{vmatrix} 3 & 5 \\ 2 & 4 \end{vmatrix} + 0 \cdot \begin{vmatrix} 3 & 2 \\ 2 & 1 \end{vmatrix}
$$

$$
= 1 \cdot \left( 2 \cdot 4 - 5 \cdot 1 \right) = 1 \cdot (8 - 5) = 1 \cdot 3 = 3
$$

#### Fourth determinant:
$$
\begin{vmatrix}
1 & 0 & 0 \\
3 & 2 & 1 \\
2 & 1 & 4
\end{vmatrix}
= 1 \cdot \begin{vmatrix} 2 & 1 \\ 1 & 4 \end{vmatrix} - 0 \cdot \begin{vmatrix} 3 & 1 \\ 2 & 4 \end{vmatrix} + 0 \cdot \begin{vmatrix} 3 & 2 \\ 2 & 1 \end{vmatrix}
$$

$$
= 1 \cdot \left( 2 \cdot 4 - 1 \cdot 1 \right) = 1 \cdot (8 - 1) = 1 \cdot 7 = 7
$$

### **Final calculation:**

Now substitute these values back into the cofactor expansion:

$$
\text{det}(\mathbf{C}) = 2 \cdot 42 - 3 \cdot 40 + 1 \cdot 3 - 4 \cdot 7
$$

$$
= 84 - 120 + 3 - 28
$$

$$
= 84 - 120 + 3 - 28 = -61
$$

Thus, the determinant of \(\mathbf{C}\) is:

$$
\boxed{\text{det}(\mathbf{C}) = -61}
$$

### Step 2: Calculating the 4x4 Minors

#### First Minor:
$$
\begin{vmatrix} 4 & 0 & 0 & 7 \\ 0 & 0 & 0 & 0 \\ 1 & 4 & 3 & 2 \\ 2 & 3 & 4 & 5 \end{vmatrix}
$$

Notice that the second row has all zeros, so the determinant of this matrix is \(0\):

$$
\text{det} = 0
$$

#### Second Minor:
$$
\begin{vmatrix} 1 & 0 & 0 & 7 \\ 3 & 0 & 0 & 0 \\ 1 & 4 & 3 & 2 \\ 2 & 3 & 4 & 5 \end{vmatrix}
$$

We will expand along the first column:

$$
\text{det} = 1 \cdot \begin{vmatrix} 0 & 0 & 0 \\ 4 & 3 & 2 \\ 3 & 4 & 5 \end{vmatrix} - 0 \cdot \text{det} + 0 \cdot \text{det} + 7 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 1 & 4 & 3 \\ 2 & 3 & 4 \end{vmatrix}
$$

The first minor will be zero because of the first row being \(0\), so we just calculate the last term:

$$
7 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 1 & 4 & 3 \\ 2 & 3 & 4 \end{vmatrix}
$$

Expanding along the first column:

$$
= 7 \cdot 3 \cdot \begin{vmatrix} 4 & 3 \\ 3 & 4 \end{vmatrix} = 7 \cdot 3 \cdot (4 \cdot 4 - 3 \cdot 3)
$$

$$
= 7 \cdot 3 \cdot (16 - 9) = 7 \cdot 3 \cdot 7 = 147
$$

Thus, the second minor is \(147\).

#### Third Minor:
$$
\begin{vmatrix} 1 & 4 & 0 & 7 \\ 3 & 0 & 0 & 0 \\ 2 & 3 & 4 & 5 \\ 1 & 2 & 3 & 4 \end{vmatrix}
$$

We will expand along the first column:

$$
\text{det} = 1 \cdot \begin{vmatrix} 0 & 0 & 0 \\ 3 & 4 & 5 \\ 2 & 3 & 4 \end{vmatrix} - 4 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 2 & 4 & 5 \\ 1 & 2 & 4 \end{vmatrix} + 0 \cdot \text{det} + 7 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 2 & 3 & 4 \\ 1 & 2 & 3 \end{vmatrix}
$$

The first and third minors will be zero due to the presence of zeros in the rows, so we just calculate the second term:

$$
-4 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 2 & 4 & 5 \\ 1 & 2 & 4 \end{vmatrix}
$$

Expanding along the first column:

$$
= -4 \cdot 3 \cdot \begin{vmatrix} 4 & 5 \\ 2 & 4 \end{vmatrix} = -4 \cdot 3 \cdot (4 \cdot 4 - 5 \cdot 2)
$$

$$
= -4 \cdot 3 \cdot (16 - 10) = -4 \cdot 3 \cdot 6 = -72
$$

Thus, the third minor is \(-72\).

#### Fourth Minor:
$$
\begin{vmatrix} 1 & 4 & 0 & 0 \\ 3 & 0 & 0 & 0 \\ 2 & 3 & 4 & 5 \\ 1 & 2 & 3 & 4 \end{vmatrix}
$$

Expanding along the first column:

$$
\text{det} = 1 \cdot \begin{vmatrix} 0 & 0 & 0 \\ 3 & 4 & 5 \\ 2 & 3 & 4 \end{vmatrix} - 4 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 2 & 4 & 5 \\ 1 & 2 & 4 \end{vmatrix}

The first term is zero, and the second term will give:

$$
-4 \cdot \begin{vmatrix} 3 & 0 & 0 \\ 2 & 4 & 5 \\ 1 & 2 & 4 \end{vmatrix}

We already calculated this minor as \(-72\), so:

$$
\text{det} = -4 \cdot (-72) = 288

#### Fifth Minor:
$$
\begin{vmatrix} 1 & 4 & 0 & 0 \\ 3 & 0 & 0 & 0 \\ 2 & 1 & 4 & 3 \\ 1 & 2 & 3 & 4 \end{vmatrix}

Expanding along the first column:

The first term is zero, and the second term will give:

This minor is computed as:

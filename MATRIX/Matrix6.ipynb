## Exercises for Students

### Perform row and column operations to reduce the following matrices to an upper triangular form and calculate their determinants by taking the product of the diagonal elements.

---

### 1. Matrix \( A \):

$$
A = \begin{bmatrix}
12 & 3 \\
-18 & -4
\end{bmatrix}
$$

#### Solution:

1. **Perform row operations:**

- Multiply row 1 by \( \frac{1}{12} \) to normalize the first element:

$$ R_1 \rightarrow \frac{1}{12} R_1 $$

$$
A = \begin{bmatrix}
1 & \frac{1}{4} \\
-18 & -4
\end{bmatrix}
$$

- Add \( 18 \) times row 1 to row 2 to eliminate the first element of row 2:

$$ R_2 \rightarrow R_2 + 18R_1 $$

$$
A = \begin{bmatrix}
1 & \frac{1}{4} \\
0 & \frac{1}{2}
\end{bmatrix}
$$

2. **Upper triangular form:**

The matrix is now in upper triangular form:

$$
A = \begin{bmatrix}
1 & \frac{1}{4} \\
0 & \frac{1}{2}
\end{bmatrix}
$$

3. **Calculate the determinant:**

The determinant of an upper triangular matrix is the product of the diagonal elements:

$$ \text{det}(A) = 1 \times \frac{1}{2} = \frac{1}{2} $$

---

### 2. Matrix \( B \):

$$
B = \begin{bmatrix} 
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9 
\end{bmatrix}
$$

#### Solution:

1. **Perform row operations:**

- Subtract 4 times row 1 from row 2:

$$ R_2 \rightarrow R_2 - 4R_1 $$

$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
7 & 8 & 9
\end{bmatrix}
$$

- Subtract 7 times row 1 from row 3:

$$ R_3 \rightarrow R_3 - 7R_1 $$

$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
0 & -6 & -12
\end{bmatrix}
$$

- Divide row 2 by \( -3 \) to make the pivot element 1:

$$ R_2 \rightarrow \frac{1}{-3} R_2 $$

$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & 1 & 2 \\
0 & -6 & -12
\end{bmatrix}
$$

- Add 6 times row 2 to row 3 to eliminate the second element of row 3:

$$ R_3 \rightarrow R_3 + 6R_2 $$

$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & 1 & 2 \\
0 & 0 & 0
\end{bmatrix}
$$

2. **Upper triangular form:**

The matrix is now in upper triangular form:

$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & 1 & 2 \\
0 & 0 & 0
\end{bmatrix}
$$

3. **Calculate the determinant:**

The determinant of an upper triangular matrix is the product of the diagonal elements:

$$ \text{det}(B) = 1 \times 1 \times 0 = 0 $$

---

### Final Answer:

- The determinant of \( A \) is \( \frac{1}{2} \).
- The determinant of \( B \) is \( 0 \).

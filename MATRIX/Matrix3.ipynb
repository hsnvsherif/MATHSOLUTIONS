## Exercises for Students

### 1. 2x2 Matrices:

$$
\mathbf{A} =
\begin{pmatrix}
2 & 3 \\
1 & 4
\end{pmatrix}
, \qquad
\mathbf{B} =
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix}
, \qquad
\mathbf{C} =
\begin{pmatrix}
-1 & 2 \\
3 & 0
\end{pmatrix}
$$

**Solution for Determinants of 2x2 Matrices:**

1. $\text{det}(\mathbf{A})$:
$$
\text{det}(\mathbf{A}) = 2 \cdot 4 - 3 \cdot 1 = 8 - 3 = 5
$$

2. $\text{det}(\mathbf{B})$:
$$
\text{det}(\mathbf{B}) = 5 \cdot 8 - 6 \cdot 7 = 40 - 42 = -2
$$

3. $\text{det}(\mathbf{C})$:
$$
\text{det}(\mathbf{C}) = (-1) \cdot 0 - 2 \cdot 3 = 0 - 6 = -6
$$

### 2. 3x3 Matrices:

$$
\mathbf{D} =
\begin{pmatrix}
1 & 0 & 2 \\
-1 & 3 & 1 \\
2 & 4 & -2
\end{pmatrix}
, \qquad
\mathbf{E} =
\begin{pmatrix}
3 & 1 & -1 \\
0 & 2 & 4 \\
5 & 3 & 2
\end{pmatrix}
, \qquad
\mathbf{F} =
\begin{pmatrix}
2 & -3 & 1 \\
1 & 4 & -2 \\
1 & 5 & 3
\end{pmatrix}
$$

**Solution for Determinants of 3x3 Matrices:**

1. $\text{det}(\mathbf{D})$:

Using cofactor expansion along the first row:
$$
\text{det}(\mathbf{D}) = 1 \cdot \begin{vmatrix} 3 & 1 \\ 4 & -2 \end{vmatrix} - 0 \cdot \begin{vmatrix} -1 & 1 \\ 2 & -2 \end{vmatrix} + 2 \cdot \begin{vmatrix} -1 & 3 \\ 2 & 4 \end{vmatrix}
$$
Calculating the minors:
$$
\begin{vmatrix} 3 & 1 \\ 4 & -2 \end{vmatrix} = 3 \cdot (-2) - 1 \cdot 4 = -6 - 4 = -10
$$
$$
\begin{vmatrix} -1 & 3 \\ 2 & 4 \end{vmatrix} = (-1) \cdot 4 - 3 \cdot 2 = -4 - 6 = -10
$$
Now, substituting back into the determinant:
$$
\text{det}(\mathbf{D}) = 1 \cdot (-10) - 0 + 2 \cdot (-10) = -10 + (-20) = -30
$$

2. $\text{det}(\mathbf{E})$:

Using cofactor expansion along the first row:
$$
\text{det}(\mathbf{E}) = 3 \cdot \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} - 1 \cdot \begin{vmatrix} 0 & 4 \\ 5 & 2 \end{vmatrix} + (-1) \cdot \begin{vmatrix} 0 & 2 \\ 5 & 3 \end{vmatrix}
$$
Calculating the minors:
$$
\begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} = 2 \cdot 2 - 4 \cdot 3 = 4 - 12 = -8
$$
$$
\begin{vmatrix} 0 & 4 \\ 5 & 2 \end{vmatrix} = 0 \cdot 2 - 4 \cdot 5 = 0 - 20 = -20
$$
$$
\begin{vmatrix} 0 & 2 \\ 5 & 3 \end{vmatrix} = 0 \cdot 3 - 2 \cdot 5 = 0 - 10 = -10
$$
Now, substituting back into the determinant:
$$
\text{det}(\mathbf{E}) = 3 \cdot (-8) - 1 \cdot (-20) + (-1) \cdot (-10) = -24 + 20 + 10 = 6
$$

3. $\text{det}(\mathbf{F})$:

Using cofactor expansion along the first row:
$$
\text{det}(\mathbf{F}) = 2 \cdot \begin{vmatrix} 4 & -2 \\ 5 & 3 \end{vmatrix} - (-3) \cdot \begin{vmatrix} 1 & -2 \\ 1 & 3 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 4 \\ 1 & 5 \end{vmatrix}
$$
Calculating the minors:
$$
\begin{vmatrix} 4 & -2 \\ 5 & 3 \end{vmatrix} = 4 \cdot 3 - (-2) \cdot 5 = 12 + 10 = 22
$$
$$
\begin{vmatrix} 1 & -2 \\ 1 & 3 \end{vmatrix} = 1 \cdot 3 - (-2) \cdot 1 = 3 + 2 = 5
$$
$$
\begin{vmatrix} 1 & 4 \\ 1 & 5 \end{vmatrix} = 1 \cdot 5 - 4 \cdot 1 = 5 - 4 = 1
$$
Now, substituting back into the determinant:
$$
\text{det}(\mathbf{F}) = 2 \cdot 22 + 3 \cdot 5 + 1 \cdot 1 = 44 + 15 + 1 = 60
$$

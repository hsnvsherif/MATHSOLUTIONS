### 1. Find the inverse matrix for

$$
\begin{bmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{bmatrix}
$$

**Solution for Inverse Matrix:**

To find the inverse of a matrix, we use the formula:
$$
\mathbf{A}^{-1} = \frac{1}{\text{det}(\mathbf{A})} \text{adj}(\mathbf{A})
$$
where $\text{det}(\mathbf{A})$ is the determinant and $\text{adj}(\mathbf{A})$ is the adjugate matrix.

For matrix $\mathbf{A}$:
$$
\mathbf{A} =
\begin{bmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{bmatrix}
$$

---


1. **Determinant**: 
   The determinant of $\mathbf{A}$ is calculated as:
   $$
   \text{det}(\mathbf{A}) = 2 \cdot \begin{vmatrix} 1 & 0 \\ 2 & 0 \end{vmatrix} - 0 \cdot \begin{vmatrix} 0 & 0 \\ 1 & 0 \end{vmatrix} + 1 \cdot \begin{vmatrix} 0 & 1 \\ 1 & 2 \end{vmatrix}
   $$

   First minor:
   $$
   \begin{vmatrix} 1 & 0 \\ 2 & 0 \end{vmatrix} = (1 \cdot 0 - 0 \cdot 2) = 0
   $$

   Second minor:
   $$
   \begin{vmatrix} 0 & 1 \\ 1 & 2 \end{vmatrix} = (0 \cdot 2 - 1 \cdot 1) = -1
   $$

   Now, substituting into the determinant formula:
   $$
   \text{det}(\mathbf{A}) = 2 \cdot 0 + 0 + 1 \cdot (-1) = -1
   $$


---


2. **Adjugate Matrix**:
   We now find the adjugate of $\mathbf{A}$, which involves calculating the cofactors of each element in $\mathbf{A}$ and then transposing the matrix of cofactors.

   After calculating the cofactors, the adjugate matrix is:
   $$
   \text{adj}(\mathbf{A}) =
   \begin{bmatrix}
   -2 & 0 & 1 \\
   0 & 0 & 0 \\
   1 & -1 & 2
   \end{bmatrix}
   $$

---

3. **Inverse Matrix**:
   Using the formula for the inverse:
   $$
   \mathbf{A}^{-1} = \frac{1}{\text{det}(\mathbf{A})} \cdot \text{adj}(\mathbf{A}) = \frac{1}{-1} \cdot
   \begin{bmatrix}
   -2 & 0 & 1 \\
   0 & 0 & 0 \\
   1 & -1 & 2
   \end{bmatrix}
   $$

   So, the inverse matrix is:
   $$
   \mathbf{A}^{-1} =
   \begin{bmatrix}
   2 & 0 & -1 \\
   0 & 0 & 0 \\
   -1 & 1 & -2
   \end{bmatrix}
   $$

### 2. Using the `det()` command, determine the rank of the matrix below

$$
A = 
\begin{bmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{bmatrix}
$$

**Solution for Rank of Matrix:**

---


To determine the rank of the matrix, we use the `det()` command to check the determinant of the matrix. If the determinant is non-zero, the matrix has full rank (i.e., rank = 3). If the determinant is zero, we check for smaller minors to determine the rank.

1. **Determinant of matrix $A$**:
   $$
   \text{det}(A) = 4 \cdot \begin{vmatrix} 6 & 3 \\ 9 & 1 \end{vmatrix} - (-3) \cdot \begin{vmatrix} -1 & 3 \\ 2 & 1 \end{vmatrix} + 7 \cdot \begin{vmatrix} -1 & 6 \\ 2 & 9 \end{vmatrix}
   $$

   First minor:
   $$
   \begin{vmatrix} 6 & 3 \\ 9 & 1 \end{vmatrix} = 6 \cdot 1 - 3 \cdot 9 = 6 - 27 = -21
   $$

   Second minor:
   $$
   \begin{vmatrix} -1 & 3 \\ 2 & 1 \end{vmatrix} = (-1) \cdot 1 - 3 \cdot 2 = -1 - 6 = -7
   $$

   Third minor:
   $$
   \begin{vmatrix} -1 & 6 \\ 2 & 9 \end{vmatrix} = (-1) \cdot 9 - 6 \cdot 2 = -9 - 12 = -21
   $$

   Now, substituting into the determinant formula:
   $$
   \text{det}(A) = 4 \cdot (-21) + 3 \cdot (-7) + 7 \cdot (-21) = -84 - 21 - 147 = -252
   $$

Since the determinant is non-zero ($\text{det}(A) = -252$), matrix $A$ has full rank, so the rank is 3.

## First-Order Ordinary Differential Equations

Solve the following first-order ordinary differential equations:

1. 
   $$ y'(x) = y(x) $$

   **Solution:**  
   This is a separable differential equation:  
   $$ \frac{dy}{dx} = y $$  
   Separating variables:  
   $$ \frac{dy}{y} = dx $$  
   Integrating both sides:  
   $$ \ln |y| = x + C $$  
   Solving for \( y \):  
   $$ y = C e^x $$  
   where \( C \) is the constant of integration.

---

2.  
   $$ y'(x) = \frac{1}{2y(x)} $$

   **Solution:**  
   Again, this is a separable equation:  
   $$ 2y \frac{dy}{dx} = 1 $$  
   Rearranging:  
   $$ 2y dy = dx $$  
   Integrating both sides:  
   $$ \int 2y dy = \int dx $$  
   $$ y^2 = x + C $$  
   Solving for \( y \):  
   $$ y = \pm \sqrt{x + C} $$  
   where \( C \) is the constant of integration.

## Second-Order Ordinary Differential Equation

Solve the second-order ordinary differential equation:

$$
y''(x) + y'(x) = 0
$$

with boundary conditions:

$$
y(0) = 2, \quad y'(0) = -1
$$

---

### **Solution**
#### **Step 1: Solve the Characteristic Equation**
The corresponding characteristic equation is:

$$
r^2 + r = 0
$$

Factorizing:

$$
r (r + 1) = 0
$$

Thus, the roots are:

$$
r_1 = 0, \quad r_2 = -1
$$

Since we have distinct real roots, the general solution is:

$$
y(x) = C_1 + C_2 e^{-x}
$$

---

### **Step 2: Apply Boundary Conditions**
#### **First Condition: \( y(0) = 2 \)**
Substituting \( x = 0 \):

$$
\begin{aligned}
y(0) &= C_1 + C_2 e^{0} \\
2 &= C_1 + C_2
\end{aligned}
$$

---

#### **Second Condition: \( y'(0) = -1 \)**
First, differentiate \( y(x) \):

$$
y'(x) = -C_2 e^{-x}
$$

Substituting \( x = 0 \):

$$
\begin{aligned}
y'(0) &= -C_2 e^{0} \\
-1 &= -C_2
\end{aligned}
$$

Solving for \( C_2 \):

$$
C_2 = 1
$$

---

### **Step 3: Solve for \( C_1 \)**
$$
\begin{aligned}
C_1 + C_2 &= 2 \\
C_1 + 1 &= 2 \\
C_1 &= 1
\end{aligned}
$$

---

### **Final Solution**
$$
y(x) = 1 + e^{-x}
$$

## Second-Order Ordinary Differential Equation

Solve the second-order ordinary differential equation:

$$
y''(x) - y(x) = 0
$$

with boundary conditions:

$$
y(0) = 2, \quad y'(0) = 0
$$

---

### **Solution**
#### **Step 1: Solve the Characteristic Equation**
The characteristic equation is:

$$
r^2 - 1 = 0
$$

Factorizing:

$$
(r - 1)(r + 1) = 0
$$

Thus, the roots are:

$$
r_1 = 1, \quad r_2 = -1
$$

Since the roots are real and distinct, the general solution is:

$$
y(x) = C_1 e^x + C_2 e^{-x}
$$

---

### **Step 2: Apply Boundary Conditions**
#### **First Condition: \( y(0) = 2 \)**
Substituting \( x = 0 \):

$$
\begin{aligned}
y(0) &= C_1 e^0 + C_2 e^0 \\
2 &= C_1 + C_2
\end{aligned}
$$

---

#### **Second Condition: \( y'(0) = 0 \)**
First, differentiate \( y(x) \):

$$
y'(x) = C_1 e^x - C_2 e^{-x}
$$

Substituting \( x = 0 \):

$$
\begin{aligned}
y'(0) &= C_1 e^0 - C_2 e^0 \\
0 &= C_1 - C_2
\end{aligned}
$$

---

### **Step 3: Solve for \( C_1 \) and \( C_2 \)**
Solving the system of equations:

$$
\begin{aligned}
C_1 + C_2 &= 2 \\
C_1 - C_2 &= 0
\end{aligned}
$$

Adding both equations:

$$
2C_1 = 2
$$

$$
C_1 = 1
$$

Substituting \( C_1 = 1 \) into \( C_1 + C_2 = 2 \):

$$
1 + C_2 = 2
$$

$$
C_2 = 1
$$

---

### **Final Solution**
$$
y(x) = e^x + e^{-x}
$$

or equivalently:

$$
y(x) = 2 \cosh(x)
$$

since \( \cosh(x) = \frac{e^x + e^{-x}}{2} \).

## Second-Order Ordinary Differential Equation

Solve the second-order ordinary differential equation:

$$
\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)
$$

---

### **Solution**
#### **Step 1: Solve the Characteristic Equation**
Rewriting the equation:

$$
y''(x) + \omega^2 y(x) = 0
$$

The corresponding **characteristic equation** is:

$$
r^2 + \omega^2 = 0
$$

Solving for \( r \):

$$
r^2 = -\omega^2
$$

$$
r = \pm i\omega
$$

---

### **Step 2: Write the General Solution**
Since the characteristic roots are **purely imaginary**, the general solution is given by:

$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

where \( C_1 \) and \( C_2 \) are constants determined by initial or boundary conditions.

---

### **Final Solution**
$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

This represents a **harmonic oscillation** with angular frequency \( \omega \).

5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

## Verifying the Wave Equation

Check if the function:

$$
\psi(t, x) = A \cos(\omega t + kx)
$$

is a solution of the **wave equation**:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

where the wave speed is given by:

$$
v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}.
$$

---

### **Step 1: Compute the Second Partial Derivatives**
#### **First, the Second Time Derivative**
Computing the first derivative with respect to \( t \):

$$
\frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx)
$$

Taking the second derivative:

$$
\frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$$

---

#### **Second, the Second Spatial Derivative**
Computing the first derivative with respect to \( x \):

$$
\frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx)
$$

Taking the second derivative:

$$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$$

---

### **Step 2: Verify the Wave Equation**
The given wave equation is:

$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0.
$$

Substituting the computed second derivatives:

$$
- A \omega^2 \cos(\omega t + kx) - v^2 (- A k^2 \cos(\omega t + kx)) = 0.
$$

Since \( v^2 = \frac{\omega^2}{k^2} \), we substitute:

$$
- A \omega^2 \cos(\omega t + kx) + \frac{\omega^2}{k^2} A k^2 \cos(\omega t + kx) = 0.
$$

Simplifying:

$$
- A \omega^2 \cos(\omega t + kx) + A \omega^2 \cos(\omega t + kx) = 0.
$$

Since both terms cancel out, the equation holds true.

---

### **Conclusion**
Thus, the function:

$$
\psi(t, x) = A \cos(\omega t + kx)
$$

**is a valid solution of the wave equation**:

$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0.
$$

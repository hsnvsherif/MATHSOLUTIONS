3. Calculate the area of the region bounded by the lines:
$x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$.

### **Area of the Bounded Region**

The given boundaries are:
- **Vertical limits:** \( x = 1 \) and \( x = 2 \)
- **Horizontal limit:** \( y = 0 \) (x-axis)
- **Upper function:** \( y = x^2 + 1 \)

The area is given by the definite integral:

\[
A = \int_{1}^{2} (x^2 + 1) \,dx
\]

#### **Step-by-Step Solution**

Evaluating the integral:

\[
\begin{align*}
A &= \int_{1}^{2} (x^2 + 1) \,dx \\[8pt]
&= \int_{1}^{2} x^2 \,dx + \int_{1}^{2} 1 \,dx
\end{align*}
\]

Computing each integral separately:

\[
\begin{align*}
\int x^2 \,dx &= \frac{x^3}{3} \\[8pt]
\int 1 \,dx &= x
\end{align*}
\]

Applying the limits \( 1 \) to \( 2 \):

\[
\begin{align*}
A &= \left[ \frac{x^3}{3} \right]_{1}^{2} + \left[ x \right]_{1}^{2} \\[8pt]
&= \left( \frac{2^3}{3} - \frac{1^3}{3} \right) + (2 - 1) \\[8pt]
&= \left( \frac{8}{3} - \frac{1}{3} \right) + 1 \\[8pt]
&= \frac{7}{3} + 1 \\[8pt]
&= \frac{7}{3} + \frac{3}{3} \\[8pt]
&= \frac{10}{3}
\end{align*}
\]

4. Calculate the area under the sine curve over the interval $[0, \pi]$, using:

$$P = \int_a^b f(x) \, dx = \int_0^\pi \sin(x) \, dx$$

### **Area Under the Sine Curve**

We calculate the definite integral:

\[
P = \int_{0}^{\pi} \sin(x) \,dx
\]

#### **Step-by-Step Solution**

The integral of \( \sin(x) \) is:

\[
\int \sin(x) \,dx = -\cos(x) + C
\]

Applying the limits \( 0 \) to \( \pi \):

\[
\begin{align*}
P &= \left[ -\cos(x) \right]_{0}^{\pi} \\[8pt]
&= \left( -\cos(\pi) \right) - \left( -\cos(0) \right) \\[8pt]
&= \left( -(-1) \right) - \left( -(1) \right) \\[8pt]
&= (1) - (-1) \\[8pt]
&= 1 + 1 \\[8pt]
&= 2
\end{align*}
\]

Thus, the **area under the sine curve** over \( [0, \pi] \) is:

\[
P = 2
\]

5. Calculate the length of the sine curve over the same interval using:

$$L = \int_a^b \sqrt{1 + (f'(x))^2} \, dx= \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx
$$ 

# Length of the Sine Curve Over \( [0, \pi] \)

The length of a curve \( y = f(x) \) from \( x = a \) to \( x = b \) is given by:

$$
L = \int_a^b \sqrt{1 + (f'(x))^2} \,dx
$$

For \( f(x) = \sin(x) \), we first compute its derivative:

$$
f'(x) = \cos(x)
$$

Thus, the arc length integral becomes:

$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx
$$

## **Exact Integral Expression**
The integral:

$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx
$$

**does not have an elementary closed-form solution** and is typically evaluated using numerical methods.

## **Approximating the Value**
Using numerical integration, we find:

$$
L \approx 3.8202
$$

Thus, the **length of the sine curve over** \( [0, \pi] \) **is approximately**:

$$
L \approx 3.82
$$

6. Find the distance of the moving particle between time $t=0$ and $t=2$ for the following position function: $x(t) = 3t^2 - 6t + 1$.

### **Distance of the Moving Particle**

We are given the position function:

\[
x(t) = 3t^2 - 6t + 1
\]

We need to find the **total distance traveled** from \( t = 0 \) to \( t = 2 \). The total distance is given by:

\[
D = \int_{0}^{2} |v(t)| \, dt
\]

where \( v(t) \) is the velocity function:

\[
v(t) = \frac{dx}{dt} = x'(t)
\]

### **1️⃣ Compute the Velocity Function**

\[
\begin{align*}
v(t) &= \frac{d}{dt} (3t^2 - 6t + 1) \\[8pt]
     &= 6t - 6
\end{align*}
\]

### **2️⃣ Find Critical Points**
Setting \( v(t) = 0 \):

\[
\begin{align*}
6t - 6 &= 0 \\[8pt]
t &= 1
\end{align*}
\]

Since \( v(t) \) changes sign at \( t = 1 \), we split the integral at this point.

### **3️⃣ Compute the Total Distance**

\[
D = \int_0^1 |v(t)| \, dt + \int_1^2 |v(t)| \, dt
\]

For \( 0 \leq t < 1 \), \( v(t) < 0 \), so:

\[
|v(t)| = -(6t - 6) = -6t + 6
\]

For \( 1 \leq t \leq 2 \), \( v(t) > 0 \), so:

\[
|v(t)| = 6t - 6
\]

\[
D = \int_0^1 (-6t + 6) \, dt + \int_1^2 (6t - 6) \, dt
\]

### **4️⃣ Evaluate Each Integral**
First integral:

\[
\begin{align*}
\int_0^1 (-6t + 6) \, dt &= \left[ -3t^2 + 6t \right]_{0}^{1} \\[8pt]
&= (-3(1)^2 + 6(1)) - (-3(0)^2 + 6(0)) \\[8pt]
&= (-3 + 6) - (0) \\[8pt]
&= 3
\end{align*}
\]

Second integral:

\[
\begin{align*}
\int_1^2 (6t - 6) \, dt &= \left[ 3t^2 - 6t \right]_{1}^{2} \\[8pt]
&= (3(2)^2 - 6(2)) - (3(1)^2 - 6(1)) \\[8pt]
&= (3(4) - 12) - (3(1) - 6) \\[8pt]
&= (12 - 12) - (3 - 6) \\[8pt]
&= 0 + 3 \\[8pt]
&= 3
\end{align*}
\]

### **5️⃣ Compute Total Distance**

\[
\begin{align*}
D &= 3 + 3 \\[8pt]
D &= 6
\end{align*}
\]

Thus the answer is : 6

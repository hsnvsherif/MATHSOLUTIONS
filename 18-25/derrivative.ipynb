2. Prove:
   - $\frac{d}{dx} (\ln(\sin(x))) = \cot(x)$

# **2. Proof of \( \frac{d}{dx} (\ln(\sin(x))) = \cot(x) \)**

We need to differentiate the function:

$$
y = \ln(\sin(x)).
$$

---

### **Step 1: Apply the Chain Rule**
Using the derivative formula for \( \ln(f(x)) \):

$$
\frac{d}{dx} \ln(f(x)) = \frac{1}{f(x)} \cdot f'(x),
$$

where \( f(x) = \sin(x) \), we differentiate:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x).
$$

---

### **Step 2: Simplify the Expression**
Since we know that:

$$
\frac{\cos(x)}{\sin(x)} = \cot(x),
$$

we obtain:

$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x).
$$

---

### **Final Conclusion:**
Thus, we have proven that:

$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x).
$$

3. For $f(x) = \cos(x)$, verify that $f''(x) = -f(x)$.

# **3. Verification of \( f''(x) = -f(x) \) for \( f(x) = \cos(x) \)**

We need to verify that for \( f(x) = \cos(x) \):

$$
f''(x) = -f(x).
$$

---

### **Step 1: Compute the First Derivative**
Differentiating \( f(x) = \cos(x) \), we get:

$$
f'(x) = \frac{d}{dx} \cos(x) = -\sin(x).
$$

---

### **Step 2: Compute the Second Derivative**
Differentiating again:

$$
f''(x) = \frac{d}{dx} (-\sin(x)) = -\cos(x).
$$

---

### **Step 3: Compare with \( -f(x) \)**
Since we know that:

$$
f(x) = \cos(x),
$$

multiplying by \( -1 \):

$$
- f(x) = -\cos(x).
$$

Thus, we confirm:

$$
f''(x) = -f(x).
$$

---

### **Final Conclusion:**
Since we have shown that:

$$
f''(x) = -\cos(x) = -f(x),
$$

we have verified that:

$$
f''(x) = -f(x).
$$

4. Using de l'Hospital's Rule, find the improper limits:
   - $\displaystyle \lim_{x\to 0} \frac{\sin{x}}{x}$

We need to evaluate the limit:

$$
\lim_{x\to 0} \frac{\sin{x}}{x}.
$$

---

### **Step 1: Check the Indeterminate Form**
Substituting \( x = 0 \):

$$
\frac{\sin(0)}{0} = \frac{0}{0},
$$

which is an **indeterminate form**. Since we have a \( \frac{0}{0} \) case, we apply **L'Hôpital's Rule**.

---

### **Step 2: Apply L'Hôpital's Rule**
L'Hôpital's Rule states that if a limit results in \( \frac{0}{0} \) or \( \frac{\infty}{\infty} \), we differentiate the numerator and denominator separately:

$$
\lim_{x\to 0} \frac{\sin{x}}{x} = \lim_{x\to 0} \frac{\cos{x}}{1}.
$$

---

### **Step 3: Evaluate the New Limit**
Since \( \cos(0) = 1 \), we get:

$$
\lim_{x\to 0} \frac{\cos{x}}{1} = 1.
$$

---

### **Final Conclusion:**
Thus, using **L'Hôpital's Rule**, we have shown that:

$$
\lim_{x\to 0} \frac{\sin{x}}{x} = 1.
$$

   - $\displaystyle \lim_{x\to \infty} \frac{\ln x}{x}$

We need to evaluate the limit:

$$
\lim_{x\to \infty} \frac{\ln x}{x}.
$$

---

### **Step 1: Check the Indeterminate Form**
As \( x \to \infty \):

- \( \ln x \to \infty \)
- \( x \to \infty \)

Thus, we get the indeterminate form:

$$
\frac{\infty}{\infty}.
$$

Since this is an **indeterminate form**, we apply **L'Hôpital's Rule**.

---

### **Step 2: Apply L'Hôpital's Rule**
L'Hôpital's Rule states that if a limit results in \( \frac{\infty}{\infty} \), we differentiate the numerator and denominator separately:

$$
\lim_{x\to \infty} \frac{\ln x}{x} = \lim_{x\to \infty} \frac{\frac{d}{dx} (\ln x)}{\frac{d}{dx} (x)}.
$$

Differentiating:

- \( \frac{d}{dx} (\ln x) = \frac{1}{x} \)
- \( \frac{d}{dx} (x) = 1 \)

Thus, we obtain:

$$
\lim_{x\to \infty} \frac{\frac{1}{x}}{1} = \lim_{x\to \infty} \frac{1}{x}.
$$

---

### **Step 3: Evaluate the New Limit**
Since \( \frac{1}{x} \to 0 \) as \( x \to \infty \), we conclude:

$$
\lim_{x\to \infty} \frac{1}{x} = 0.
$$

---

### **Final Conclusion:**
Thus, using **L'Hôpital's Rule**, we have shown that:

$$
\lim_{x\to \infty} \frac{\ln x}{x} = 0.
$$

5. In physics, the position of a particle is given by $x(t) = 3t^2 - 6t + 1$. Find the velocity $V(t)=x'(t)$ and acceleration $a(t)=V'(t)=x''(t)$ of the particle at time $t = 2$.

# **5. Finding Velocity and Acceleration of a Particle**

The position of the particle is given by:

$$
x(t) = 3t^2 - 6t + 1.
$$

We need to find:

- **Velocity**: \( V(t) = x'(t) \)
- **Acceleration**: \( a(t) = V'(t) = x''(t) \)
- Evaluate both at \( t = 2 \).

---

### **Step 1: Compute the Velocity \( V(t) \)**
Velocity is the first derivative of position:

$$
V(t) = \frac{d}{dt} (3t^2 - 6t + 1).
$$

Using the power rule:

$$
V(t) = 6t - 6.
$$

At \( t = 2 \):

$$
V(2) = 6(2) - 6 = 12 - 6 = 6.
$$

Thus, the **velocity at \( t = 2 \) is**:

$$
V(2) = 6.
$$

---

### **Step 2: Compute the Acceleration \( a(t) \)**
Acceleration is the derivative of velocity:

$$
a(t) = \frac{d}{dt} (6t - 6).
$$

Since the derivative of \( 6t \) is 6 and the derivative of \( -6 \) is 0:

$$
a(t) = 6.
$$

At \( t = 2 \):

$$
a(2) = 6.
$$

Thus, the **acceleration at \( t = 2 \) is**:

$$
a(2) = 6.
$$

---

### **Final Conclusion:**
- The **velocity** of the particle at \( t = 2 \) is \( V(2) = 6 \).
- The **acceleration** of the particle at \( t = 2 \) is \( a(2) = 6 \).

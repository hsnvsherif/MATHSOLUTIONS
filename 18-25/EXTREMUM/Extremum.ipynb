6. The profit function is $P(u) = -2u^2 + 50u - 300$, where $u$ is the number of units sold. Find the number of units that maximize profit.

# **6. Finding the Number of Units that Maximize Profit**

The profit function is given by:

$$
P(u) = -2u^2 + 50u - 300.
$$

where \( u \) is the number of units sold. We need to find the value of \( u \) that **maximizes profit**.

---

### **Step 1: Find the Critical Points**
To find the maximum profit, we take the derivative of \( P(u) \) and set it to zero:

$$
P'(u) = \frac{d}{du} (-2u^2 + 50u - 300).
$$

Using the power rule:

$$
P'(u) = -4u + 50.
$$

Setting \( P'(u) = 0 \) to find critical points:

$$
-4u + 50 = 0.
$$

Solving for \( u \):

$$
4u = 50 \quad \Rightarrow \quad u = \frac{50}{4} = 12.5.
$$

Since the number of units must be a whole number, we consider \( u = 12 \) or \( u = 13 \).

---

### **Step 2: Verify Maximum using Second Derivative**
Taking the second derivative:

$$
P''(u) = \frac{d}{du} (-4u + 50) = -4.
$$

Since \( P''(u) = -4 \) is **negative**, the function is concave downward, confirming that \( u = 12.5 \) is a **maximum**.

---

### **Final Conclusion:**
Since the number of units sold must be an integer, the **optimal number of units that maximizes profit is either**:

$$
u = 12 \quad \text{or} \quad u = 13.
$$

7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangular. Find the dimensions of the rectangle.

# **7. Maximizing the Area of a Rectangle with 10 Meters of String**

We have **10 meters of string** to enclose the **largest possible rectangle**. We need to find the **dimensions** that maximize the **enclosed area**.

---

### **Step 1: Define Variables**
Let:
- \( x \) = **length** of the rectangle
- \( y \) = **width** of the rectangle
- The **total perimeter is 10 meters**, so:

$$
2x + 2y = 10.
$$

Simplifying:

$$
x + y = 5.
$$

Solving for \( y \):

$$
y = 5 - x.
$$

---

### **Step 2: Define the Area Function**
The **area** of the rectangle is:

$$
A = x \cdot y.
$$

Substituting \( y = 5 - x \):

$$
A(x) = x(5 - x).
$$

Expanding:

$$
A(x) = 5x - x^2.
$$

---

### **Step 3: Maximize the Area**
To find the maximum area, we take the **derivative**:

$$
A'(x) = \frac{d}{dx} (5x - x^2).
$$

Using the power rule:

$$
A'(x) = 5 - 2x.
$$

Setting \( A'(x) = 0 \) to find the critical points:

$$
5 - 2x = 0.
$$

Solving for \( x \):

$$
2x = 5 \quad \Rightarrow \quad x = \frac{5}{2} = 2.5.
$$

Since \( y = 5 - x \), we get:

$$
y = 5 - 2.5 = 2.5.
$$

---

### **Step 4: Confirm Maximum Using Second Derivative**
The **second derivative** is:

$$
A''(x) = -2.
$$

Since \( A''(x) < 0 \), the function is **concave downward**, confirming a **maximum**.

---

### **Final Conclusion:**
The **dimensions of the rectangle that enclose the maximum area** are:

$$
x = 2.5 \text{ meters}, \quad y = 2.5 \text{ meters}.
$$

Thus, the **optimal shape is a square** of **\( 2.5 \times 2.5 \) meters**.

8. Find extremum od $f(x) = x^2 + 3x - 5$.

# **8. Finding the Extremum of \( f(x) = x^2 + 3x - 5 \)**

We need to find the **extremum (minimum or maximum)** of the function:

$$
f(x) = x^2 + 3x - 5.
$$

---

### **Step 1: Compute the First Derivative**
To find the critical points, we take the first derivative:

$$
f'(x) = \frac{d}{dx} (x^2 + 3x - 5).
$$

Using the power rule:

$$
f'(x) = 2x + 3.
$$

Setting \( f'(x) = 0 \) to find critical points:

$$
2x + 3 = 0.
$$

Solving for \( x \):

$$
x = -\frac{3}{2}.
$$

---

### **Step 2: Verify the Nature of the Extremum**
We take the **second derivative**:

$$
f''(x) = \frac{d}{dx} (2x + 3).
$$

Since:

$$
f''(x) = 2 > 0,
$$

this means \( f(x) \) is **concave upward**, confirming a **minimum** at \( x = -\frac{3}{2} \).

---

### **Step 3: Find the Minimum Value**
To find \( f(x) \) at the critical point:

$$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5.
$$

Expanding:

$$
\frac{9}{4} - \frac{9}{2} - 5.
$$

Converting to a common denominator:

$$
\frac{9}{4} - \frac{18}{4} - \frac{20}{4} = \frac{9 - 18 - 20}{4} = \frac{-29}{4}.
$$

Thus, the **minimum value** is:

$$
f\left(-\frac{3}{2}\right) = \frac{-29}{4}.
$$

---

### **Final Conclusion:**
- **The function has a minimum at** \( x = -\frac{3}{2} \).
- **The minimum value of the function is** \( f(x) = \frac{-29}{4} \).

9. Find extremum of $f(x) =\frac{x^2+2x+1}{x-1}$.

# **9. Finding the Extremum of \( f(x) = \frac{x^2+2x+1}{x-1} \)**

We need to find the **extremum** of the function:

$$
f(x) = \frac{x^2+2x+1}{x-1}.
$$

---

### **Step 1: Compute the First Derivative**
We use the **quotient rule**, where:

- \( g(x) = x^2 + 2x + 1 \)
- \( h(x) = x - 1 \)

The quotient rule states:

$$
f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{h(x)^2}.
$$

First, compute derivatives:

$$
g'(x) = 2x + 2, \quad h'(x) = 1.
$$

Applying the quotient rule:

$$
f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}.
$$

Expanding:

$$
f'(x) = \frac{(2x^2 - 2x + 2x - 2) - (x^2 + 2x + 1)}{(x - 1)^2}.
$$

Simplifying:

$$
f'(x) = \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2}.
$$

$$
f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}.
$$

Factor the numerator:

$$
f'(x) = \frac{(x - 3)(x + 1)}{(x - 1)^2}.
$$

---

### **Step 2: Find Critical Points**
Set \( f'(x) = 0 \):

$$
(x - 3)(x + 1) = 0.
$$

Solving for \( x \):

$$
x = 3, \quad x = -1.
$$

**Check domain restrictions:** Since the denominator contains \( (x - 1)^2 \), we note that \( x = 1 \) is undefined.

Thus, the **critical points** are:

$$
x = -1, \quad x = 3.
$$

---

### **Step 3: Second Derivative Test**
To determine whether these critical points are **minima or maxima**, we compute the **second derivative**:

Differentiate \( f'(x) \):

$$
f''(x) = \frac{(x - 1)^2(2x - 2) - (x^2 - 2x - 3)(2(x - 1))}{(x - 1)^4}.
$$

At \( x = -1 \):

$$
f''(-1) = \frac{(-2)(4)}{4} = -2 < 0.
$$

Since \( f''(-1) < 0 \), **\( x = -1 \) is a local maximum**.

At \( x = 3 \):

$$
f''(3) = \frac{(2)(4)}{4} = 2 > 0.
$$

Since \( f''(3) > 0 \), **\( x = 3 \) is a local minimum**.

---

### **Step 4: Find Function Values at Extremum**
Evaluate \( f(x) \):

$$
f(-1) = \frac{(-1)^2 + 2(-1) + 1}{-1 - 1} = \frac{1 - 2 + 1}{-2} = \frac{0}{-2} = 0.
$$

$$
f(3) = \frac{3^2 + 2(3) + 1}{3 - 1} = \frac{9 + 6 + 1}{2} = \frac{16}{2} = 8.
$$

---

### **Final Conclusion:**
- **Local Maximum at** \( x = -1 \), with **\( f(-1) = 0 \)**.
- **Local Minimum at** \( x = 3 \), with **\( f(3) = 8 \)**.
- **The function is undefined at** \( x = 1 \) (vertical asymptote).

   - $f(g(x))$
   - $g(f(x))$
   - $f(f(x))$
   - $g(g(x))$

## Function Composition

Given the functions:

$$
\begin{aligned}
f(x) &= 3x - 1 \\
g(x) &= \sqrt{x}
\end{aligned}
$$

### Finding Compositions:

#### 1. \( f(g(x)) \)

$$
\begin{aligned}
f(g(x)) &= f(\sqrt{x}) \\
&= 3\sqrt{x} - 1
\end{aligned}
$$

#### 2. \( g(f(x)) \)

$$
\begin{aligned}
g(f(x)) &= g(3x - 1) \\
&= \sqrt{3x - 1}
\end{aligned}
$$

#### 3. \( f(f(x)) \)

$$
\begin{aligned}
f(f(x)) &= f(3x - 1) \\
&= 3(3x - 1) - 1 \\
&= 9x - 3 - 1 \\
&= 9x - 4
\end{aligned}
$$

#### 4. \( g(g(x)) \)

$$
\begin{aligned}
g(g(x)) &= g(\sqrt{x}) \\
&= \sqrt{\sqrt{x}} \\
&= x^{\frac{1}{4}}
\end{aligned}
$$

## Function Composition Check

Given the functions:

$$
\begin{aligned}
f(x) &= e^x \\
g(x) &= \ln(x)
\end{aligned}
$$

### Checking Compositions:

#### 1. \( f(g(x)) \)

$$
\begin{aligned}
f(g(x)) &= f(\ln(x)) \\
&= e^{\ln(x)} \\
&= x
\end{aligned}
$$

#### 2. \( g(f(x)) \)

$$
\begin{aligned}
g(f(x)) &= g(e^x) \\
&= \ln(e^x) \\
&= x
\end{aligned}
$$

### Conclusion:

We observe that:

$$
f(g(x)) = x \quad \text{and} \quad g(f(x)) = x
$$

This means that the functions **\( f(x) = e^x \) and \( g(x) = \ln(x) \) are inverses of each other.** Applying one function and then the other returns the original input \( x \).

4. We have function $f=\{(1,7), (2,9), (3,11)\}$. Give inverse function $f^{-1}$.

## Finding the Inverse of a Function

Given the function:

$$
f = \{(1,7), (2,9), (3,11)\}
$$

To find the inverse function \( f^{-1} \), we swap each \( (x, y) \) pair to \( (y, x) \):

$$
f^{-1} = \{(7,1), (9,2), (11,3)\}
$$

### **Conclusion:**
The inverse function \( f^{-1} \) consists of the ordered pairs:

$$
f^{-1} = \{(7,1), (9,2), (11,3)\}
$$



5. We have function $f=\{(1,7), (2,7), (3,11)\}$. Give inverse function $f^{-1}$.

## Finding the Inverse of a Function

Given the function:

$$
f = \{(1,7), (2,7), (3,11)\}
$$

To find the inverse function \( f^{-1} \), we swap each \( (x, y) \) pair to \( (y, x) \):

$$
f^{-1} = \{(7,1), (7,2), (11,3)\}
$$

### **Important Observation:**
The inverse relation \( f^{-1} \) is:

$$
\{(7,1), (7,2), (11,3)\}
$$

Since the value **7 maps to both 1 and 2**, the inverse **is not a function** because a function must have only one output for each input.

### **Conclusion:**
- The inverse relation exists, but it **is not a function** because one input (7) has multiple outputs (1 and 2).
- For a function to have an inverse that is also a function, it must be **one-to-one (bijective)**, meaning each \( x \) must map to a unique \( y \).

 We have function $f(x)= x-1$. Give inverse function $f^{-1}$. Show both functions on the same Geogebra notebook.

## Finding the Inverse of \( f(x) = x - 1 \)

To find the inverse function \( f^{-1}(x) \), we follow these steps:

1. Start with the given function:

   $$
   y = x - 1
   $$

2. Swap \( x \) and \( y \):

   $$
   x = y - 1
   $$

3. Solve for \( y \):

   $$
   \begin{aligned}
   x + 1 &= y \\
   y &= x + 1
   \end{aligned}
   $$

Thus, the inverse function is:

$$
f^{-1}(x) = x + 1
$$

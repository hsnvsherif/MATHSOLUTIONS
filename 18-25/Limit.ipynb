 Calculate:
   - $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$

## Calculating the Limit

We are given the limit:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

### **Step 1: Divide by the Highest Power of \( n \)**
The highest power of \( n \) in the fraction is \( n^2 \), so we divide both the numerator and denominator by \( n^2 \):

$$
\begin{aligned}
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} &= \lim_{n \to \infty} \frac{\frac{n^2}{n^2} + \frac{3n}{n^2}}{\frac{2n^2}{n^2} - \frac{2n}{n^2}} \\
&= \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
\end{aligned}
$$

### **Step 2: Evaluate the Limit**
As \( n \to \infty \), the terms \( \frac{3}{n} \) and \( \frac{2}{n} \) approach 0, so we get:

$$
\frac{1 + 0}{2 - 0} = \frac{1}{2}
$$

### **Final Answer:**
$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}
$$

   - $\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}$

## Calculating the Limit

We are given the limit:

$$
\lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}
$$

### **Step 1: Expand the Numerator**
Expanding \( (2n+3)^3 \) using the binomial theorem:

$$
(2n+3)^3 = 8n^3 + 12n^2 \cdot 3 + 6n \cdot 9 + 27 = 8n^3 + 36n^2 + 54n + 27
$$

Thus, the given limit becomes:

$$
\lim_{n \to \infty} \frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1}
$$

### **Step 2: Divide by the Highest Power of \( n \)**
The highest power of \( n \) in the fraction is \( n^3 \), so we divide both the numerator and denominator by \( n^3 \):

$$
\begin{aligned}
\lim_{n \to \infty} \frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1} &= \lim_{n \to \infty} \frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}
\end{aligned}
$$

### **Step 3: Evaluate the Limit**
As \( n \to \infty \), the terms \( \frac{36}{n}, \frac{54}{n^2}, \) and \( \frac{27}{n^3} \) approach **0**, and \( \frac{1}{n^3} \) also approaches **0**. Therefore, we get:

$$
\frac{8 + 0 + 0 + 0}{1 - 0} = \frac{8}{1} = 8
$$

### **Final Answer:**
$$
\lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1} = 8
$$

2. Prove using the squeeze theorem:
   - $\displaystyle\lim_{n \to \infty} \frac{\sin(n)}{n}$

## Proving the Limit Using the Squeeze Theorem

We need to prove:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

### **Step 1: Establish Bounds for \( \sin(n) \)**
We know that the sine function is always bounded by:

$$
-1 \leq \sin(n) \leq 1
$$

Dividing by \( n \) (which is positive for large \( n \)), we get:

$$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

### **Step 2: Compute the Limits of the Bounds**
Now, we compute the limits of the two bounding functions:

$$
\lim_{n \to \infty} \frac{-1}{n} = 0, \quad \text{and} \quad \lim_{n \to \infty} \frac{1}{n} = 0.
$$

Since both the upper and lower bounds approach **0**, the **Squeeze Theorem** states that:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0.
$$

### **Final Conclusion:**
Using the **Squeeze Theorem**, we conclude:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0.
$$

4. Find the limit of the sequence:
   - $a_n = (1+\frac{1}{n})^n$

## Finding the Limit of the Sequence

We are given the sequence:

$$
a_n = \left(1 + \frac{1}{n}\right)^n
$$

### **Step 1: Recognizing the Limit Definition**
From calculus, we know the fundamental limit:

$$
\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e
$$

where \( e \) is the **Euler's number**, approximately **2.718**.

### **Step 2: Applying the Limit**
Taking the limit as \( n \to \infty \), we directly obtain:

$$
\lim_{n \to \infty} a_n = e
$$

### **Final Conclusion:**
The sequence \( a_n \) converges to:

$$
\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e
$$

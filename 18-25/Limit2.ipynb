1. Compute:
   - $\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

## Computing the Limit

We are given the limit:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

### **Step 1: Divide by the Highest Power of \( x \)**
The highest power of \( x \) in the denominator is \( x^4 \). We divide both the numerator and denominator by \( x^4 \):

$$
\begin{aligned}
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} &= \lim_{x \to \infty} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}} \\
&= \lim_{x \to \infty} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}
\end{aligned}
$$

### **Step 2: Evaluate the Limit**
As \( x \to \infty \), the terms \( \frac{1}{x} \) and \( \frac{2}{x^2} \) approach **0**, and \( \frac{3}{x} \) also approaches **0**. Therefore, we get:

$$
\frac{0 + 0}{1 - 0} = \frac{0}{1} = 0.
$$

### **Final Answer:**
$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$$

2. Find:
   
   - $\displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1}$.


## Computing the Limit

We need to evaluate the limit:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x+1}
$$

### **Step 1: Apply Direct Substitution**
Substituting \( x = 0 \):

$$
\frac{\sin(3(0))}{2(0) + 1} = \frac{\sin(0)}{1} = \frac{0}{1} = 0.
$$

### **Step 2: Confirm Using the Small-Angle Approximation**
For small \( x \), we use the approximation:

$$
\sin(3x) \approx 3x \quad \text{as} \quad x \to 0.
$$

Substituting this approximation:

$$
\lim_{x \to 0} \frac{3x}{2x+1}.
$$

### **Step 3: Evaluate the Limit Using Algebra**
Dividing both numerator and denominator by \( x \):

$$
\lim_{x \to 0} \frac{3}{2 + \frac{1}{x}}.
$$

As \( x \to 0 \), the fraction \( \frac{1}{x} \) grows large, making the denominator approach **1**. Thus, we confirm:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x+1} = 0.
$$

### **Final Answer:**
$$
\lim_{x \to 0} \frac{\sin(3x)}{2x+1} = 0.
$$

4. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$


## 2. Asymptotes of    - $f(x) = \frac{x^2 - 1}{x^2 + 1}$

### **Step 1: Find Vertical Asymptotes**
Vertical asymptotes occur where the denominator is **zero**, but since:

$$
x^2 + 1 > 0 \quad \text{for all real } x
$$

the denominator never equals zero. Thus, there are **no vertical asymptotes**.

---

### **Step 2: Find Horizontal Asymptotes**
We check the limit as \( x \to \pm\infty \):

$$
\lim_{x \to \infty} \frac{\sin(x)}{x^2+1}
$$

Since \( -1 \leq \sin(x) \leq 1 \), we apply the **Squeeze Theorem**:

$$
- \frac{1}{x^2+1} \leq \frac{\sin(x)}{x^2+1} \leq \frac{1}{x^2+1}
$$

As \( x \to \infty \), both \( \frac{1}{x^2+1} \) and \( -\frac{1}{x^2+1} \) approach **0**, so by the **Squeeze Theorem**:

$$
\lim_{x \to \infty} g(x) = 0
$$

Thus, \( y = 0 \) is a **horizontal asymptote**.



### **Step 3: Check for Oblique Asymptotes**
Since the numerator \( \sin(x) \) is **bounded**, and the denominator \( x^2 + 1 \) grows indefinitely, the function approaches 0 smoothly, meaning **no oblique asymptotes**.

# Taylor Series Expansion of $f(x) = \cos(x)$ at $x = 0$

## 1. Taylor Series Definition
The Taylor series of a function $f(x)$ centered at $x = a$ is given by:

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x - a)^n
$$

For $f(x) = \cos(x)$, we expand around $x = 0$.

## 2. Compute the Taylor Series Terms
The derivatives of $f(x) = \cos(x)$ are:

$$
\begin{aligned}
f(x) &= \cos(x), \\
f'(x) &= -\sin(x), \\
f''(x) &= -\cos(x), \\
f'''(x) &= \sin(x), \\
f''''(x) &= \cos(x)
\end{aligned}
$$

Evaluating at $x = 0$:

$$
\begin{aligned}
f(0) &= 1, \\
f'(0) &= 0, \\
f''(0) &= -1, \\
f'''(0) &= 0, \\
f''''(0) &= 1
\end{aligned}
$$

Thus, the Taylor series up to the 4th degree is:

$$
\cos(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!}
$$

which simplifies to:

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

## 3. Conclusion
The 4th-degree Taylor approximation of $\cos(x)$ around $x = 0$ is:

$$
T_4(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

# Taylor Series Expansion of $h(x) = \frac{1}{1 - x}$ at $x = 0$

## 1. Taylor Series Definition
The Taylor series of a function $f(x)$ centered at $x = a$ is given by:

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x - a)^n
$$

For $h(x) = \frac{1}{1 - x}$, we expand around $x = 0$.

## 2. Compute the Taylor Series Terms

### **First few derivatives:**
$$
\begin{aligned}
h(x) &= (1 - x)^{-1}, \\
h'(x) &= (1 - x)^{-2}, \\
h''(x) &= 2(1 - x)^{-3}, \\
h'''(x) &= 6(1 - x)^{-4}, \\
h''''(x) &= 24(1 - x)^{-5}.
\end{aligned}
$$

### **Evaluate at $x = 0$:**
$$
\begin{aligned}
h(0) &= 1, \\
h'(0) &= 1, \\
h''(0) &= 2, \\
h'''(0) &= 6, \\
h''''(0) &= 24.
\end{aligned}
$$

Using the Taylor series formula:

$$
h(x) \approx h(0) + h'(0) x + \frac{h''(0)}{2!} x^2 + \frac{h'''(0)}{3!} x^3 + \frac{h''''(0)}{4!} x^4
$$

## 3. Substituting Values
$$
\begin{aligned}
h(x) &\approx 1 + x + \frac{2}{2!} x^2 + \frac{6}{3!} x^3 + \frac{24}{4!} x^4, \\
&\approx 1 + x + x^2 + x^3 + x^4.
\end{aligned}
$$

Thus, the **4th-degree Taylor series** of $h(x) = \frac{1}{1 - x}$ around $x = 0$ is:

$$
h(x) \approx 1 + x + x^2 + x^3 + x^4.
$$

This series is a well-known **geometric series expansion**:

$$
\frac{1}{1 - x} = \sum_{n=0}^{\infty} x^n, \quad \text{for } |x| < 1.
$$

# Taylor Series Expansion of $g(x) = \sin(x)$ at $x = \pi$

## 1. Taylor Series Definition
The Taylor series of a function $f(x)$ centered at $x = a$ is given by:

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x - a)^n
$$

For $g(x) = \sin(x)$, we expand around $x = \pi$.

## 2. Compute the Taylor Series Terms

### **First few derivatives:**
$$
\begin{aligned}
g(x) &= \sin(x), \\
g'(x) &= \cos(x), \\
g''(x) &= -\sin(x), \\
g'''(x) &= -\cos(x), \\
g''''(x) &= \sin(x).
\end{aligned}
$$

### **Evaluate at $x = \pi$:**
$$
\begin{aligned}
g(\pi) &= \sin(\pi) = 0, \\
g'(\pi) &= \cos(\pi) = -1, \\
g''(\pi) &= -\sin(\pi) = 0, \\
g'''(\pi) &= -\cos(\pi) = 1, \\
g''''(\pi) &= \sin(\pi) = 0.
\end{aligned}
$$

Using the Taylor series formula:

$$
g(x) \approx g(\pi) + g'(\pi) (x - \pi) + \frac{g''(\pi)}{2!} (x - \pi)^2 + \frac{g'''(\pi)}{3!} (x - \pi)^3 + \frac{g''''(\pi)}{4!} (x - \pi)^4
$$

## 3. Substituting Values
$$
\begin{aligned}
g(x) &\approx 0 - (x - \pi) + \frac{0}{2!} (x - \pi)^2 + \frac{1}{3!} (x - \pi)^3 + \frac{0}{4!} (x - \pi)^4, \\
&\approx -(x - \pi) + \frac{(x - \pi)^3}{6}.
\end{aligned}
$$

Thus, the **4th-degree Taylor series** of $g(x) = \sin(x)$ around $x = \pi$ is:

$$
g(x) \approx -(x - \pi) + \frac{(x - \pi)^3}{6}.
$$

# Finding the Tangent Line to $f(x) = e^{\sin(x)}$ at $x_0 = \pi$

## 1. Tangent Line Formula
The equation of a tangent line to a function $f(x)$ at $x = x_0$ is given by:

$$
y = f'(x_0) (x - x_0) + f(x_0)
$$

where:
- $f(x)$ is the function,
- $f'(x_0)$ is the derivative evaluated at $x_0$.

## 2. Compute $f(x)$ and $f'(x)$

Given:
$$
f(x) = e^{\sin(x)}
$$

### **Derivative of $f(x)$**
Using the chain rule:

$$
f'(x) = e^{\sin(x)} \cdot \cos(x)
$$

### **Evaluate at $x_0 = \pi$**
Since:

$$
\sin(\pi) = 0, \quad \cos(\pi) = -1
$$

we get:

$$
\begin{aligned}
f(\pi) &= e^{\sin(\pi)} = e^0 = 1, \\
f'(\pi) &= e^{\sin(\pi)} \cdot \cos(\pi) = e^0 \cdot (-1) = -1.
\end{aligned}
$$

## 3. Tangent Line Equation
Substituting into the tangent line formula:

$$
y = (-1)(x - \pi) + 1
$$

which simplifies to:

$$
y = -x + \pi + 1.
$$

Thus, the equation of the **tangent line** to $f(x) = e^{\sin(x)}$ at $x_0 = \pi$ is:

$$
y = -x + \pi + 1.
$$

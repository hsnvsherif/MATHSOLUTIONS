Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?


**No, the spheres with equations x² + y² + z² = 1 and x² + y² + z² = 2 do not have any common points.**

Here's why:

* **Interpretation:**
    * The equation x² + y² + z² = 1 represents a sphere centered at the origin (0, 0, 0) with a radius of 1.
    * The equation x² + y² + z² = 2 represents a sphere centered at the origin (0, 0, 0) with a radius of √2.

* **Analysis:**
    * The first sphere has a radius of 1, meaning all its points are exactly 1 unit away from the origin.
    * The second sphere has a radius of √2 (approximately 1.41), meaning all its points are exactly √2 units away from the origin.

* **Conclusion:**
    * Since the radii are different, there is no point that satisfies both equations simultaneously. In other words, no point can be both 1 unit and √2 units away from the origin.

**Therefore, the two spheres do not intersect.**

 What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x-1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.


**1. Find the intersection:**

To find the intersection of the two spheres, we need to solve the system of equations:

  $$
  \begin{align*}
    x^2 + y^2 + z^2 &= 1 \\
    (x-1)^2 + y^2 + z^2 &= 1
  \end{align*}
  $$

**2. Simplify the equations:**

* Expand the second equation: 
   $$
   x^2 - 2x + 1 + y^2 + z^2 = 1
   $$

* Rearrange the second equation:
   $$
   x^2 - 2x + y^2 + z^2 = 0
   $$

* Subtract the first equation from the rearranged second equation:
   $$
   -2x + 1 = 0 
   $$

* Solve for x:
   $$
   x = \frac{1}{2}
   $$

**3. Substitute the value of x into either of the sphere equations:**

* Substitute x = 1/2 into the first equation:
   $$
   \left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1
   $$
   $$
   \frac{1}{4} + y^2 + z^2 = 1
   $$
   $$
   y^2 + z^2 = \frac{3}{4}
   $$

**4. Identify the curve:**

* The equation y² + z² = 3/4 represents a circle in the yz-plane. 

* Since x is constant (x = 1/2), the intersection of the two spheres forms a **circle** in a plane parallel to the yz-plane.

**Therefore, the intersection of the two spheres is a circle lying in the plane x = 1/2.**

* Write the equation of the tangent plane to the paraboloid $z=(x-1)^2+y^2+1$ at point $P=(1,0,1)$.
**1. Find the Gradient of the Function**

* Define the function: 
   $$
   f(x, y) = (x - 1)^2 + y^2 + 1 
   $$ 
* Calculate the partial derivatives:
   $$
   \frac{\partial f}{\partial x} = 2(x - 1) 
   $$
   $$
   \frac{\partial f}{\partial y} = 2y 
   $$ 
* The gradient of f(x, y) is:
   $$
   \nabla f(x, y) = \left\langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right\rangle = \langle 2(x - 1), 2y \rangle 
   $$

**2. Evaluate the Gradient at the Point P(1, 0)**

* Substitute the coordinates of point P into the gradient:
   $$
   \nabla f(1, 0) = \langle 2(1 - 1), 2(0) \rangle = \langle 0, 0 \rangle 
   $$

**3. Determine the Equation of the Tangent Plane**

* The equation of the tangent plane at point P(x₀, y₀, z₀) is given by:
   $$
   z - z₀ = \frac{\partial f}{\partial x}(x₀, y₀)(x - x₀) + \frac{\partial f}{\partial y}(x₀, y₀)(y - y₀) 
   $$

* Substitute the values:
   $$
   z - 1 = 0(x - 1) + 0(y - 0) 
   $$
   $$
   z - 1 = 0 
   $$

**Therefore, the equation of the tangent plane to the paraboloid at point P(1, 0, 1) is:**

$$
z = 1

### Equation of a Circle

1. **General Equation of a Circle**:
   The general equation of a circle with center \( (h, k) \) and radius \( r \) is given by:
   $$
   (x - h)^2 + (y - k)^2 = r^2
   $$

2. **Substitute the Given Values**:
   Here, the center is \( A(1, 2) \), so \( h = 1 \) and \( k = 2 \), and the radius is \( r = 3 \). Substituting these values into the general equation:
   $$
   (x - 1)^2 + (y - 2)^2 = 3^2
   $$

3. **Simplify the Equation**:
   Simplifying the equation:
   $$
   (x - 1)^2 + (y - 2)^2 = 9
   $$

Thus, the equation of the circle is:
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

### Equation of a Parabola

1. **General Form of the Parabola**:
   The equation of a parabola that intersects the \( O_x \)-axis at points \( x = 2 \) and \( x = 4 \) is of the form:
   $$
   y = a(x - 2)(x - 4)
   $$

2. **Substitute the Known Point**:
   The parabola passes through the point \( (3, 1) \), so substitute \( x = 3 \) and \( y = 1 \) into the equation to find \( a \):
   $$
   1 = a(3 - 2)(3 - 4)
   $$

3. **Solve for \( a \)**:
   Simplifying the equation:
   $$
   1 = a(1)(-1)
   $$
   $$
   1 = -a
   $$
   So, \( a = -1 \).

4. **Final Equation of the Parabola**:
   Substituting \( a = -1 \) into the equation:
   $$
   y = -1(x - 2)(x - 4)
   $$

Thus, the equation of the parabola is:
$$
y = -(x - 2)(x - 4)
$$

### Find the Center of the Ellipse

We are given the equation of the ellipse:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

1. **Rearrange the equation**:
   Group the \( x \)-terms and \( y \)-terms:
   $$
   (x^2 - 4x) + 4(y^2 - 4y) = -16
   $$

2. **Complete the square** for the \( x \)-terms and \( y \)-terms:

   - For \( x^2 - 4x \), the square of half of \( -4 \) is \( (-2)^2 = 4 \), so add and subtract 4.
   - For \( y^2 - 4y \), the square of half of \( -4 \) is \( (-2)^2 = 4 \), so add and subtract 4, but remember to multiply by 4 because of the coefficient 4 in front of \( y^2 \).

   So, the equation becomes:
   $$
   (x^2 - 4x + 4) + 4(y^2 - 4y + 4) = -16 + 4 + 16
   $$

3. **Simplify the equation**:
   $$
   (x - 2)^2 + 4(y - 2)^2 = 4
   $$

4. **Rewrite the equation in standard form**:
   Divide through by 4 to get:
   $$
   \frac{(x - 2)^2}{4} + (y - 2)^2 = 1
   $$

5. **Identify the center**:
   The center of the ellipse is given by the values of \( h \) and \( k \) in the standard form \( \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1 \). In this case, the center is \( (h, k) = (2, 2) \).

### Final Answer:
The center of the ellipse is \( (2, 2) \).

### Problem: Find the slope \( m \) of the line \( y = mx - 5 \) that is tangent to the circle \( x^2 + y^2 = 1 \).

1. **Equation of the Circle**:
   The equation of the circle is:
   $$
   x^2 + y^2 = 1
   $$

2. **Equation of the Line**:
   The equation of the line is:
   $$
   y = mx - 5
   $$

3. **Substitute the equation of the line into the equation of the circle**:
   Since the line is tangent to the circle, it touches the circle at exactly one point. We substitute \( y = mx - 5 \) into the circle's equation:
   $$
   x^2 + (mx - 5)^2 = 1
   $$

4. **Expand the equation**:
   Expanding \( (mx - 5)^2 \) gives:
   $$
   x^2 + (m^2x^2 - 10mx + 25) = 1
   $$

5. **Simplify the equation**:
   Combine like terms:
   $$
   (1 + m^2)x^2 - 10mx + 25 = 1
   $$
   Simplifying further:
   $$
   (1 + m^2)x^2 - 10mx + 24 = 0
   $$

6. **Condition for tangency**:
   For the line to be tangent to the circle, the quadratic equation must have exactly one solution. This occurs when the discriminant is zero. The discriminant \( \Delta \) of a quadratic equation \( ax^2 + bx + c = 0 \) is given by:
   $$
   \Delta = b^2 - 4ac
   $$

   For the quadratic equation \( (1 + m^2)x^2 - 10mx + 24 = 0 \), we have:
   - \( a = 1 + m^2 \)
   - \( b = -10m \)
   - \( c = 24 \)

   The discriminant is:
   $$
   \Delta = (-10m)^2 - 4(1 + m^2)(24)
   $$

7. **Solve for \( m \)**:
   Set the discriminant equal to zero:
   $$
   100m^2 - 4(1 + m^2)(24) = 0
   $$
   Simplify the equation:
   $$
   100m^2 - 96 - 96m^2 = 0
   $$
   Combine like terms:
   $$
   4m^2 = 96
   $$
   Solve for \( m \):
   $$
   m^2 = 24
   $$
   $$
   m = \sqrt{24} = 2\sqrt{6}
   $$

### Final Answer:
The slope of the tangent line is:
$$
m = 2\sqrt{6}
$$

### Problem: Find the intersection points of the hyperbola \( x^2 - y^2 = 1 \) with the ellipse \( x^2 + 4y^2 = 6 \).

1. **Equations of the Hyperbola and the Ellipse**:
   The equation of the hyperbola is:
   $$
   x^2 - y^2 = 1
   $$
   The equation of the ellipse is:
   $$
   x^2 + 4y^2 = 6
   $$

2. **Substitute the equation of the hyperbola into the equation of the ellipse**:
   To find the intersection points, substitute \( x^2 = y^2 + 1 \) from the hyperbola's equation into the ellipse's equation:
   $$
   (y^2 + 1) + 4y^2 = 6
   $$

3. **Simplify the equation**:
   Combine like terms:
   $$
   y^2 + 1 + 4y^2 = 6
   $$
   $$
   5y^2 + 1 = 6
   $$

4. **Solve for \( y^2 \)**:
   Subtract 1 from both sides:
   $$
   5y^2 = 5
   $$
   Divide by 5:
   $$
   y^2 = 1
   $$

5. **Find \( y \)**:
   Take the square root of both sides:
   $$
   y = \pm 1
   $$

6. **Find the corresponding \( x \) values**:
   Substitute \( y = 1 \) and \( y = -1 \) back into the equation of the hyperbola \( x^2 - y^2 = 1 \).

   - For \( y = 1 \):
     $$
     x^2 - 1^2 = 1 \quad \Rightarrow \quad x^2 - 1 = 1 \quad \Rightarrow \quad x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
     $$

   - For \( y = -1 \):
     $$
     x^2 - (-1)^2 = 1 \quad \Rightarrow \quad x^2 - 1 = 1 \quad \Rightarrow \quad x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
     $$

### Final Answer:
The intersection points of the hyperbola \( x^2 - y^2 = 1 \) with the ellipse \( x^2 + 4y^2 = 6 \) are:
$$
(\sqrt{2}, 1), \, (-\sqrt{2}, 1), \, (\sqrt{2}, -1), \, (-\sqrt{2}, -1)
$$

### Problem: Find the distance between the branches of the hyperbola \( x^2 - y^2 = 1 \).

1. **Equation of the Hyperbola**:
   The given equation of the hyperbola is:
   $$
   x^2 - y^2 = 1
   $$

2. **Standard Form of the Hyperbola**:
   The standard form of the equation for a hyperbola centered at the origin with horizontal transverse axis is:
   $$
   \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
   $$

   Comparing this with the given equation \( x^2 - y^2 = 1 \), we can identify \( a^2 = 1 \) and \( b^2 = 1 \). Therefore, \( a = 1 \) and \( b = 1 \).

3. **Finding the Distance Between the Branches**:
   The distance between the two branches of the hyperbola is the distance between the points on the \( x \)-axis where the hyperbola intersects. For this, we use the formula for the distance between the branches:
   $$
   \text{Distance} = 2a
   $$

   Since \( a = 1 \), the distance between the branches is:
   $$
   \text{Distance} = 2 \times 1 = 2
   $$

### Final Answer:
The distance between the branches of the hyperbola \( x^2 - y^2 = 1 \) is \( 2 \).

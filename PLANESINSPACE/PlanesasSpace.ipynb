* The plane passes through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$. Find the equation of the plane.

```python
7:1

---


from sympy import symbols, Matrix, Eq

# Define the symbols
x, y, z = symbols('x y z')

# Define the points A, B, and C
A = Matrix([1, 2, 3])
B = Matrix([3, 4, 5])
C = Matrix([2, 1, 4])

# Find two vectors in the plane
AB = B - A
AC = C - A

# Find the normal vector to the plane (cross product of AB and AC)
normal_vector = AB.cross(AC)

# Equation of the plane: n . (r - r0) = 0
# n = normal_vector, r = [x, y, z], r0 = A
plane_equation = Eq(normal_vector[0] * (x - A[0]) + normal_vector[1] * (y - A[1]) + normal_vector[2] * (z - A[2]), 0)

# Output the equation of the plane in LaTeX format with dollar signs and &
latex_code = r'''
\begin{aligned}
    \text{Normal vector:} & \quad \mathbf{n} = \mathbf{AB} \times \mathbf{AC} \\
    \text{Equation of the plane:} & \quad \mathbf{n} \cdot (\mathbf{r} - \mathbf{r}_0) = 0 \\
    \text{where} & \quad \mathbf{r} = (x, y, z) \\
    & \quad \mathbf{r}_0 = (1, 2, 3)
\end{aligned}
'''
```

We have two planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$. Find the line of intersection of these planes.

```python

from sympy import symbols, Eq, solve

# Define symbols
x, y, z, t = symbols('x y z t')

# Define the plane equations
plane1 = Eq(2*x + 3*y + 4*z, 5)
plane2 = Eq(3*x + 4*y + 2*z, 6)

# Solve the system of equations
solution = solve((plane1, plane2), (x, y, z))

# Parametric equations for the line of intersection
x_param = solution[x] + t * 1  # Adjust based on parameterized form
y_param = solution[y] + t * 1  # Adjust based on parameterized form
z_param = solution[z] + t * 1  # Adjust based on parameterized form

# Display the parametric form of the line
f"Parametric equations for the line of intersection:\n x = {x_param}, y = {y_param}, z = {z_param}"
```

### Finding Planes Parallel and Perpendicular to the Plane \( 2x + 3y + 4z = 5 \)

1. **Parallel Plane:**

   Two planes are parallel if they have the same normal vector. The equation of the plane is given as:

   $$
   2x + 3y + 4z = 5
   $$

   The normal vector to this plane is \( \mathbf{n} = (2, 3, 4) \). A parallel plane will have the same normal vector, but a different constant \( D \).

   So, the general form of the parallel plane is:

   $$
   2x + 3y + 4z = D
   $$

   We can choose any value for \( D \) other than 5. For example, let’s choose \( D = 10 \). Therefore, the equation of a parallel plane is:

   $$
   2x + 3y + 4z = 10
   $$

2. **Perpendicular Plane:**

   Two planes are perpendicular if the dot product of their normal vectors is zero. The normal vector of the plane \( 2x + 3y + 4z = 5 \) is \( \mathbf{n_1} = (2, 3, 4) \).

   To find a perpendicular plane, we need to choose a normal vector \( \mathbf{n_2} \) such that the dot product \( \mathbf{n_1} \cdot \mathbf{n_2} = 0 \).

   Let’s choose \( \mathbf{n_2} = (1, -2, 1) \). The dot product is:

   $$
   \mathbf{n_1} \cdot \mathbf{n_2} = 2(1) + 3(-2) + 4(1) = 2 - 6 + 4 = 0
   $$

   This confirms that the normal vectors are perpendicular. Therefore, the equation of the perpendicular plane can be written as:

   $$
   x - 2y + z = D
   $$

   We can choose any value for \( D \). Let’s choose \( D = 1 \). So, the equation of a perpendicular plane is:

   $$
   x - 2y + z = 1
   $$

## Finding the Distance from a Point to a Plane

The formula for the distance from a point \( A(x_0, y_0, z_0) \) to a plane \( Ax + By + Cz + D = 0 \) is:

$$
\text{Distance} = \frac{|Ax_0 + By_0 + Cz_0 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

Given the equation of the plane:

$$
2x + 3y + 4z = 5
$$

We rewrite it as:

$$
2x + 3y + 4z - 5 = 0
$$

### Step 1: Identify the components
For the plane \( 2x + 3y + 4z - 5 = 0 \), the coefficients are:
- \( A = 2 \)
- \( B = 3 \)
- \( C = 4 \)
- \( D = -5 \)

The coordinates of point \( A(1, 2, 3) \) are:
- \( x_0 = 1 \)
- \( y_0 = 2 \)
- \( z_0 = 3 \)

### Step 2: Apply the formula
Substitute the values into the distance formula:

$$
\text{Distance} = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

### Step 3: Simplify the numerator
First, simplify the expression in the numerator:

$$
2(1) + 3(2) + 4(3) - 5 = 2 + 6 + 12 - 5 = 15
$$

So, the distance becomes:

$$
\text{Distance} = \frac{|15|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

### Step 4: Simplify the denominator
Now, simplify the denominator:

$$
\sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29}
$$

### Step 5: Final answer
Thus, the distance is:

$$
\text{Distance} = \frac{15}{\sqrt{29}}
$$

* Calculate the angle between the plane $x + y + z = 1$ and the plane $x = 0$ (i.e., the $yz$ plane).

___

**1. Find the normal vectors of the planes:**

* **Plane 1:** $x + y + z = 1$ 
   The normal vector of this plane is given by the coefficients of x, y, and z: 
   $\vec{n_1} = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$

* **Plane 2:** $x = 0$ 
   The normal vector of this plane is:
   $\vec{n_2} = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}$

**2. Calculate the dot product of the normal vectors:**

$\vec{n_1} \cdot \vec{n_2} = 
\begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} \cdot 
\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} 
= 1 \times 1 + 1 \times 0 + 1 \times 0 = 1$

**3. Calculate the magnitudes of the normal vectors:**

* $|\vec{n_1}| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}$
* $|\vec{n_2}| = \sqrt{1^2 + 0^2 + 0^2} = 1$

**4. Calculate the cosine of the angle between the planes:**

The cosine of the angle θ between the planes is given by:

$\cos(\theta) = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}| \cdot |\vec{n_2}|} = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}}$

**5. Calculate the angle between the planes:**

$\theta = \arccos\left(\frac{1}{\sqrt{3}}\right) \approx 54.74^\circ$

**Therefore, the angle between the plane x + y + z = 1 and the yz plane is approximately 54.74 degrees.**
